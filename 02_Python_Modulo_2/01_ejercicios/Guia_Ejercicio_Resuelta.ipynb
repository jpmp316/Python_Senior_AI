### Ejercicio 1 — Descomposición de billetes

**Enunciado:** Diseña un programa que reciba una cantidad de dinero e indique cuántos billetes de 50 000, 20 000, 10 000, 5 000, 2 000 y 1 000 se necesitan para completar esa suma, utilizando la menor cantidad posible de billetes.

**Objetivo:** Aplicación de operadores aritméticos (división entera y módulo) y condicionales para descomposición monetaria.

```python
from dataclasses import dataclass, field
from abc import ABC, abstractmethod
from typing import List, Dict


# ============================================================
# CLASE BASE (HERENCIA + POLIMORFISMO)
# ============================================================

class EjercicioBase(ABC):
    """Clase base abstracta para todos los ejercicios."""

    @abstractmethod
    def resolver(self) -> None:
        """Cada ejercicio implementará este método."""
        pass


# ============================================================
# EJERCICIO 1 — DESCOMPOSICIÓN DE BILLETES
# ============================================================

@dataclass
class Ejercicio1(EjercicioBase):
    """
    Ejercicio 1:
    Descomponer un monto en billetes de:
    50 000, 20 000, 10 000, 5 000, 2 000 y 1 000.
    """

    _monto: int
    billetes: List[int] = field(default_factory=lambda: [50000, 20000, 10000, 5000, 2000, 1000])

    # ---------------------------------------------------------
    # ENCAPSULAMIENTO: GETTER Y SETTER
    # ---------------------------------------------------------

    @property
    def monto(self) -> int:
        return self._monto

    @monto.setter
    def monto(self, nuevo_monto: int) -> None:
        if isinstance(nuevo_monto, int) and nuevo_monto > 0:
            self._monto = nuevo_monto
        else:
            print("El monto debe ser un número entero positivo.")

    # ---------------------------------------------------------
    # LÓGICA DEL EJERCICIO
    # ---------------------------------------------------------

    def resolver(self) -> None:
        print("\n=== EJERCICIO 1: DESCOMPOSICIÓN DE BILLETES ===\n")

        restante: int = self.monto
        resultado: Dict[int, int] = {}

        for b in self.billetes:
            cantidad: int = restante // b
            resultado[b] = cantidad
            restante = restante % b

        print(f"Monto ingresado: ${self.monto:,}\n")

        for billete, cantidad in resultado.items():
            if cantidad > 0:
                print(f"Billetes de ${billete:,}: {cantidad}")

        if restante > 0:
            print(f"\nSobra una cantidad no descomponible: ${restante:,}")

        print("\nProceso completado.\n")


# ============================================================
# MAIN
# ============================================================

def main() -> None:
    try:
        monto_inicial: int = int(input("Ingrese la cantidad de dinero: "))
        ejercicio: Ejercicio1 = Ejercicio1(monto_inicial)
        ejercicio.resolver()
    except ValueError:
        print("Debe ingresar un número entero.")


if __name__ == "__main__":
    main()
```

### Ejercicio 2 — Cálculo del IMC

**Enunciado:** Crea un programa que solicite el peso (kg) y la altura (m) de una persona, valide que sean números positivos, calcule el Índice de Masa Corporal (IMC) y muestre el valor con dos decimales junto con la clasificación según los intervalos pedidos.

**Objetivo:** Validación de entradas, uso de operaciones aritméticas y condicionales para clasificación de resultados.

```python
from dataclasses import dataclass
from abc import ABC, abstractmethod


# ============================================================
# CLASE BASE (HERENCIA + POLIMORFISMO)
# ============================================================

class EjercicioBase(ABC):
    """Clase base común para todos los ejercicios."""

    @abstractmethod
    def resolver(self) -> None:
        pass


# ============================================================
# EJERCICIO 2 — CÁLCULO DEL IMC
# ============================================================

@dataclass
class Ejercicio2(EjercicioBase):
    """
    Ejercicio 2:
    Calcular el IMC validando que peso y altura sean valores positivos.
    """

    _peso: float
    _altura: float

    # ---------------------------------------------------------
    # ENCAPSULAMIENTO: GETTERS & SETTERS
    # ---------------------------------------------------------

    @property
    def peso(self) -> float:
        return self._peso

    @peso.setter
    def peso(self, nuevo_peso: float) -> None:
        if isinstance(nuevo_peso, (int, float)) and nuevo_peso > 0:
            self._peso = nuevo_peso
        else:
            print("El peso debe ser un número positivo.")

    @property
    def altura(self) -> float:
        return self._altura

    @altura.setter
    def altura(self, nueva_altura: float) -> None:
        if isinstance(nueva_altura, (int, float)) and nueva_altura > 0:
            self._altura = nueva_altura
        else:
            print("La altura debe ser un número positivo.")

    # ---------------------------------------------------------
    # LÓGICA DEL EJERCICIO
    # ---------------------------------------------------------

    def resolver(self) -> None:
        print("\n=== EJERCICIO 2: CÁLCULO DEL IMC ===\n")
        print(f"Peso ingresado: {self.peso} kg")
        print(f"Altura ingresada: {self.altura} m\n")

        imc: float = self.peso / (self.altura ** 2)

        print(f"IMC calculado: {imc:.2f}")

        # Clasificación
        if imc < 18.5:
            print("Clasificación: Bajo peso")
        elif imc < 24.9:
            print("Clasificación: Normal")
        elif imc < 29.9:
            print("Clasificación: Sobrepeso")
        else:
            print("Clasificación: Obesidad")

        print("\nProceso completado.\n")


# ============================================================
# MAIN
# ============================================================

def main() -> None:
    try:
        peso: float = float(input("Ingrese el peso en kg: "))
        altura: float = float(input("Ingrese la altura en metros: "))

        ejercicio = Ejercicio2(peso, altura)
        ejercicio.resolver()

    except ValueError:
        print("Debe ingresar valores numéricos.")


if __name__ == "__main__":
    main()
```

### Ejercicio 3 — Mayor y menor de tres números

**Enunciado:** Realiza un programa que solicite tres números al usuario y determine cuál es el mayor y cuál es el menor. Valida que las entradas sean numéricas.

**Objetivo:** Comparación múltiple con condicionales y operadores relacionales.

```python
from dataclasses import dataclass
from abc import ABC, abstractmethod


# ============================================================
# CLASE BASE (HERENCIA + POLIMORFISMO)
# ============================================================

class EjercicioBase(ABC):
    """Clase base común para todos los ejercicios."""

    @abstractmethod
    def resolver(self) -> None:
        pass


# ============================================================
# EJERCICIO 3 — MAYOR Y MENOR DE TRES NÚMEROS
# ============================================================

@dataclass
class Ejercicio3(EjercicioBase):
    """
    Ejercicio 3:
    Solicitar tres números y determinar cuál es el mayor y cuál es el menor.
    """

    _a: float
    _b: float
    _c: float

    # ---------------------------------------------------------
    # ENCAPSULAMIENTO: GETTERS & SETTERS
    # ---------------------------------------------------------

    @property
    def a(self) -> float:
        return self._a

    @a.setter
    def a(self, valor: float) -> None:
        if isinstance(valor, (int, float)):
            self._a = valor
        else:
            print("El valor debe ser un número.")

    @property
    def b(self) -> float:
        return self._b

    @b.setter
    def b(self, valor: float) -> None:
        if isinstance(valor, (int, float)):
            self._b = valor
        else:
            print("El valor debe ser un número.")

    @property
    def c(self) -> float:
        return self._c

    @c.setter
    def c(self, valor: float) -> None:
        if isinstance(valor, (int, float)):
            self._c = valor
        else:
            print("El valor debe ser un número.")

    # ---------------------------------------------------------
    # LÓGICA DEL EJERCICIO
    # ---------------------------------------------------------

    def resolver(self) -> None:
        print("\n=== EJERCICIO 3: MAYOR Y MENOR DE TRES NÚMEROS ===\n")
        print(f"Números ingresados: {self.a}, {self.b}, {self.c}\n")

        mayor = self.a
        menor = self.a

        # Evaluar mayor
        if self.b > mayor:
            mayor = self.b
        if self.c > mayor:
            mayor = self.c

        # Evaluar menor
        if self.b < menor:
            menor = self.b
        if self.c < menor:
            menor = self.c

        print(f"Mayor: {mayor}")
        print(f"Menor: {menor}")
        print("\nProceso completado.\n")


# ============================================================
# MAIN
# ============================================================

def main() -> None:
    try:
        a = float(input("Ingrese el primer número: "))
        b = float(input("Ingrese el segundo número: "))
        c = float(input("Ingrese el tercer número: "))

        ejercicio = Ejercicio3(a, b, c)
        ejercicio.resolver()

    except ValueError:
        print("Debe ingresar valores numéricos.")


if __name__ == "__main__":
    main()
```

### Ejercicio 4 — Descuento en compra

**Enunciado:** Desarrolla un programa que solicite al usuario el valor de una compra y calcule el total a pagar aplicando un descuento del 10% si el valor supera los $100000. Mostrar el valor original, el descuento aplicado y el total a pagar.

**Objetivo:** Aplicación de condicionales y operaciones aritméticas para reglas de negocio simples.

```python
from dataclasses import dataclass
from abc import ABC, abstractmethod


# ============================================================
# CLASE BASE (HERENCIA + POLIMORFISMO)
# ============================================================

class EjercicioBase(ABC):
    """Clase base común para todos los ejercicios."""

    @abstractmethod
    def resolver(self) -> None:
        pass


# ============================================================
# EJERCICIO 4 — DESCUENTO EN COMPRA
# ============================================================

@dataclass
class Ejercicio4(EjercicioBase):
    """
    Ejercicio 4:
    Calcular el total a pagar aplicando un 10% de descuento
    si la compra supera los $100000.
    """

    _compra: float

    # ---------------------------------------------------------
    # ENCAPSULAMIENTO: GETTER & SETTER
    # ---------------------------------------------------------

    @property
    def compra(self) -> float:
        return self._compra

    @compra.setter
    def compra(self, valor: float) -> None:
        if isinstance(valor, (int, float)) and valor > 0:
            self._compra = valor
        else:
            print("El valor de la compra debe ser un número positivo.")

    # ---------------------------------------------------------
    # LÓGICA DEL EJERCICIO
    # ---------------------------------------------------------

    def resolver(self) -> None:
        print("\n=== EJERCICIO 4: DESCUENTO EN COMPRA ===\n")
        print(f"Valor original de la compra: ${self.compra:,.0f}")

        descuento: float = 0.0

        if self.compra > 100000:
            descuento = self.compra * 0.10

        total_pagar: float = self.compra - descuento

        print(f"Descuento aplicado: ${descuento:,.0f}")
        print(f"Total a pagar: ${total_pagar:,.0f}")
        print("\nProceso completado.\n")


# ============================================================
# MAIN
# ============================================================

def main() -> None:
    try:
        valor = float(input("Ingrese el valor de la compra: "))
        ejercicio = Ejercicio4(valor)
        ejercicio.resolver()

    except ValueError:
        print("Debe ingresar un valor numérico válido.")


if __name__ == "__main__":
    main()
```

### Ejercicio 6 — Intento de acceso con contraseña

**Enunciado:** Escribe un programa que simule un sistema de acceso donde el usuario tiene tres intentos para ingresar una contraseña correcta. Si la acierta antes de agotar los intentos, mostrar 'Acceso concedido'; si no, 'Acceso bloqueado'.

**Objetivo:** Control de bucles con contador y validación de condiciones.

```python
from dataclasses import dataclass
from abc import ABC, abstractmethod


# ============================================================
# CLASE BASE (HERENCIA + POLIMORFISMO)
# ============================================================

class EjercicioBase(ABC):
    """Clase base común para todos los ejercicios."""

    @abstractmethod
    def resolver(self) -> None:
        pass


# ============================================================
# EJERCICIO 6 — INTENTOS DE CONTRASEÑA
# ============================================================

@dataclass
class Ejercicio6(EjercicioBase):
    """
    Ejercicio 6:
    Simular intentos de acceso con un máximo de 3 intentos.
    """

    _password_correcta: str

    # ---------------------------------------------------------
    # ENCAPSULAMIENTO: GETTER & SETTER
    # ---------------------------------------------------------

    @property
    def password_correcta(self) -> str:
        return self._password_correcta

    @password_correcta.setter
    def password_correcta(self, valor: str) -> None:
        if isinstance(valor, str) and valor.strip() != "":
            self._password_correcta = valor
        else:
            print("La contraseña debe ser un texto no vacío.")

    # ---------------------------------------------------------
    # LÓGICA DEL EJERCICIO
    # ---------------------------------------------------------

    def resolver(self) -> None:
        print("\n=== EJERCICIO 6: INTENTOS DE ACCESO ===\n")

        intentos: int = 3

        while intentos > 0:
            entrada: str = input("Ingrese la contraseña: ")

            if entrada == self.password_correcta:
                print("Acceso concedido.")
                print("\nProceso completado.\n")
                return
            else:
                intentos -= 1
                print(f"Contraseña incorrecta. Intentos restantes: {intentos}")

        print("Acceso bloqueado.")
        print("\nProceso completado.\n")


# ============================================================
# MAIN
# ============================================================

def main() -> None:
    contrasena = input("Defina la contraseña correcta para la prueba: ").strip()

    ejercicio = Ejercicio6(contrasena)
    ejercicio.resolver()


if __name__ == "__main__":
    main()
```

### Ejercicio 10 — Contar múltiplos de 3

**Enunciado:** Realiza un programa que muestre todos los números del 1 al 100 y al final indique cuántos de ellos son múltiplos de 3.

**Objetivo:** Uso de bucles con acumuladores y condicionales para conteo de propiedades numéricas.

```python
from dataclasses import dataclass
from abc import ABC, abstractmethod


# ============================================================
# CLASE BASE (HERENCIA + POLIMORFISMO)
# ============================================================

class EjercicioBase(ABC):
    """Clase base común para todos los ejercicios."""

    @abstractmethod
    def resolver(self) -> None:
        pass


# ============================================================
# EJERCICIO 10 — CONTAR MÚLTIPLOS DE 3
# ============================================================

@dataclass
class Ejercicio10(EjercicioBase):
    """
    Ejercicio 10:
    Mostrar todos los números del 1 al 100 y contar cuántos son múltiplos de 3.
    """

    _limite: int = 100

    # ---------------------------------------------------------
    # ENCAPSULAMIENTO
    # ---------------------------------------------------------

    @property
    def limite(self) -> int:
        return self._limite

    @limite.setter
    def limite(self, valor: int) -> None:
        if isinstance(valor, int) and valor > 0:
            self._limite = valor
        else:
            print("El límite debe ser un número entero positivo.")

    # ---------------------------------------------------------
    # LÓGICA DEL EJERCICIO
    # ---------------------------------------------------------

    def resolver(self) -> None:
        print("\n=== EJERCICIO 10: CONTAR MÚLTIPLOS DE 3 ===\n")
        print(f"Números del 1 al {self.limite}:\n")

        contador: int = 0

        for numero in range(1, self.limite + 1):
            print(numero)
            if numero % 3 == 0:
                contador += 1

        print(f"\nCantidad de múltiplos de 3 entre 1 y {self.limite}: {contador}")
        print("\nProceso completado.\n")


# ============================================================
# MAIN
# ============================================================

def main() -> None:
    ejercicio = Ejercicio10()  # No requiere entrada del usuario
    ejercicio.resolver()


if __name__ == "__main__":
    main()
```

**Enunciado:** Crea un programa que lea tres números diferentes y determine si pueden formar los lados de un triángulo. Si es posible, indicar si es equilátero, isósceles o escaleno. Use la desigualdad triangular para la validación.

**Objetivo:** Condicionales anidados y operadores lógicos para validar relaciones geométricas entre números.

```python
from dataclasses import dataclass
from abc import ABC, abstractmethod


# ============================================================
# CLASE BASE (HERENCIA + POLIMORFISMO)
# ============================================================

class EjercicioBase(ABC):
    """Clase base común para todos los ejercicios."""

    @abstractmethod
    def resolver(self) -> None:
        pass


# ============================================================
# EJERCICIO 11 — CLASIFICACIÓN DE TRIÁNGULO
# ============================================================

@dataclass
class Ejercicio11(EjercicioBase):
    """
    Ejercicio 11:
    Determinar si tres números pueden formar un triángulo y clasificarlo.
    """

    _a: float
    _b: float
    _c: float

    # ---------------------------------------------------------
    # ENCAPSULAMIENTO
    # ---------------------------------------------------------

    @property
    def a(self) -> float:
        return self._a

    @a.setter
    def a(self, valor: float) -> None:
        if isinstance(valor, (int, float)) and valor > 0:
            self._a = valor
        else:
            print("El valor debe ser un número positivo.")

    @property
    def b(self) -> float:
        return self._b

    @b.setter
    def b(self, valor: float) -> None:
        if isinstance(valor, (int, float)) and valor > 0:
            self._b = valor
        else:
            print("El valor debe ser un número positivo.")

    @property
    def c(self) -> float:
        return self._c

    @c.setter
    def c(self, valor: float) -> None:
        if isinstance(valor, (int, float)) and valor > 0:
            self._c = valor
        else:
            print("El valor debe ser un número positivo.")

    # ---------------------------------------------------------
    # LÓGICA DEL EJERCICIO
    # ---------------------------------------------------------

    def resolver(self) -> None:
        print("\n=== EJERCICIO 11: VALIDACIÓN Y CLASIFICACIÓN DE TRIÁNGULO ===\n")
        print(f"Lados ingresados: {self.a}, {self.b}, {self.c}\n")

        # Desigualdad triangular:
        # Cada lado debe ser menor que la suma de los otros dos.
        if (
            self.a < self.b + self.c and
            self.b < self.a + self.c and
            self.c < self.a + self.b
        ):
            print("Los valores forman un triángulo.")

            # Clasificación
            if self.a == self.b == self.c:
                print("Tipo de triángulo: Equilátero.")
            elif self.a == self.b or self.a == self.c or self.b == self.c:
                print("Tipo de triángulo: Isósceles.")
            else:
                print("Tipo de triángulo: Escaleno.")

        else:
            print("Los valores NO pueden formar un triángulo.")

        print("\nProceso completado.\n")


# ============================================================
# MAIN
# ============================================================

def main() -> None:
    try:
        a = float(input("Ingrese el primer lado: "))
        b = float(input("Ingrese el segundo lado: "))
        c = float(input("Ingrese el tercer lado: "))

        ejercicio = Ejercicio11(a, b, c)
        ejercicio.resolver()

    except ValueError:
        print("Debe ingresar valores numéricos válidos.")


if __name__ == "__main__":
    main()
```

### Ejercicio 12 — Cálculo de salario con horas extras

**Enunciado:** Desarrolla un programa que calcule el salario final de un trabajador a partir de las horas trabajadas y el valor por hora. Si trabaja más de 40 horas, las horas extras se pagan al 150% del valor normal.

**Objetivo:** Condicionales y operaciones aritméticas para cálculo de nómina básico.

```python
from dataclasses import dataclass
from abc import ABC, abstractmethod


# ============================================================
# CLASE BASE (HERENCIA + POLIMORFISMO)
# ============================================================

class EjercicioBase(ABC):
    """Clase base común para todos los ejercicios."""

    @abstractmethod
    def resolver(self) -> None:
        pass


# ============================================================
# EJERCICIO 12 — SALARIO CON HORAS EXTRAS
# ============================================================

@dataclass
class Ejercicio12(EjercicioBase):
    """
    Ejercicio 12:
    Calcular salario final considerando:
    - Pago normal por hora
    - Horas extras al 150% si exceden las 40 horas
    """

    _horas: float
    _valor_hora: float

    # ---------------------------------------------------------
    # ENCAPSULAMIENTO
    # ---------------------------------------------------------

    @property
    def horas(self) -> float:
        return self._horas

    @horas.setter
    def horas(self, valor: float) -> None:
        if isinstance(valor, (int, float)) and valor >= 0:
            self._horas = valor
        else:
            print("Las horas trabajadas deben ser un número no negativo.")

    @property
    def valor_hora(self) -> float:
        return self._valor_hora

    @valor_hora.setter
    def valor_hora(self, valor: float) -> None:
        if isinstance(valor, (int, float)) and valor > 0:
            self._valor_hora = valor
        else:
            print("El valor por hora debe ser un número positivo.")

    # ---------------------------------------------------------
    # LÓGICA DEL EJERCICIO
    # ---------------------------------------------------------

    def resolver(self) -> None:
        print("\n=== EJERCICIO 12: CÁLCULO DE SALARIO CON HORAS EXTRAS ===\n")
        print(f"Horas trabajadas: {self.horas}")
        print(f"Valor por hora: {self.valor_hora}\n")

        salario_final: float = 0.0

        if self.horas <= 40:
            salario_final = self.horas * self.valor_hora
        else:
            horas_extra = self.horas - 40
            pago_normal = 40 * self.valor_hora
            pago_extra = horas_extra * (self.valor_hora * 1.5)
            salario_final = pago_normal + pago_extra

        print(f"Salario final: ${salario_final:,.2f}")
        print("\nProceso completado.\n")


# ============================================================
# MAIN
# ============================================================

def main() -> None:
    try:
        horas = float(input("Ingrese las horas trabajadas: "))
        valor_hora = float(input("Ingrese el valor por hora: "))

        ejercicio = Ejercicio12(horas, valor_hora)
        ejercicio.resolver()

    except ValueError:
        print("Debe ingresar valores numéricos válidos.")


if __name__ == "__main__":
    main()
```

### Ejercicio 13 — Número primo

**Enunciado:** Elabora un programa que solicite un número entero positivo y determine si es primo (divisible solo por 1 y por sí mismo).

**Objetivo:** Uso de ciclos y operador módulo para comprobar divisibilidad y optimizar la búsqueda de factores.

```python
from dataclasses import dataclass
from abc import ABC, abstractmethod


# ============================================================
# CLASE BASE (HERENCIA + POLIMORFISMO)
# ============================================================

class EjercicioBase(ABC):
    """Clase base común para todos los ejercicios."""
    
    @abstractmethod
    def resolver(self) -> None:
        pass


# ============================================================
# EJERCICIO 13 — NÚMERO PRIMO
# ============================================================

@dataclass
class Ejercicio13(EjercicioBase):
    """
    Ejercicio 13:
    Determinar si un número entero positivo es primo.
    """

    _numero: int

    # ---------------------------------------------------------
    # ENCAPSULAMIENTO
    # ---------------------------------------------------------

    @property
    def numero(self) -> int:
        return self._numero

    @numero.setter
    def numero(self, valor: int) -> None:
        if isinstance(valor, int) and valor > 0:
            self._numero = valor
        else:
            print("Debe ingresar un número entero positivo.")

    # ---------------------------------------------------------
    # LÓGICA DEL EJERCICIO
    # ---------------------------------------------------------

    def resolver(self) -> None:
        print("\n=== EJERCICIO 13: NÚMERO PRIMO ===\n")
        print(f"Número ingresado: {self.numero}\n")

        if self.numero == 1:
            print("1 no se considera un número primo.")
            print("\nProceso completado.\n")
            return

        es_primo = True

        # Solo necesitamos revisar hasta la raíz cuadrada
        limite = int(self.numero ** 0.5)

        for divisor in range(2, limite + 1):
            if self.numero % divisor == 0:
                es_primo = False
                break

        if es_primo:
            print("El número es primo.")
        else:
            print("El número no es primo.")

        print("\nProceso completado.\n")


# ============================================================
# MAIN
# ============================================================

def main() -> None:
    try:
        valor = int(input("Ingrese un número entero positivo: "))
        ejercicio = Ejercicio13(valor)
        ejercicio.resolver()

    except ValueError:
        print("Debe ingresar un número entero válido.")


if __name__ == "__main__":
    main()
```

### Ejercicio 14 — Serie de Fibonacci (n términos)

**Enunciado:** Diseña un programa que muestre los primeros n términos de la serie de Fibonacci, donde n es un entero positivo ingresado por el usuario.

**Objetivo:** Manejo de ciclos y variables acumuladoras para generar secuencias recursivas iterativas.

```python
from dataclasses import dataclass
from abc import ABC, abstractmethod


# ============================================================
# CLASE BASE (HERENCIA + POLIMORFISMO)
# ============================================================

class EjercicioBase(ABC):
    """Clase base común para todos los ejercicios."""
    
    @abstractmethod
    def resolver(self) -> None:
        pass


# ============================================================
# EJERCICIO 14 — SERIE DE FIBONACCI
# ============================================================

@dataclass
class Ejercicio14(EjercicioBase):
    """
    Ejercicio 14:
    Generar los primeros n términos de la serie de Fibonacci.
    """

    _n: int

    # ---------------------------------------------------------
    # ENCAPSULAMIENTO
    # ---------------------------------------------------------

    @property
    def n(self) -> int:
        return self._n

    @n.setter
    def n(self, valor: int) -> None:
        if isinstance(valor, int) and valor > 0:
            self._n = valor
        else:
            print("Debe ingresar un número entero positivo.")

    # ---------------------------------------------------------
    # LÓGICA DEL EJERCICIO
    # ---------------------------------------------------------

    def resolver(self) -> None:
        print("\n=== EJERCICIO 14: SERIE DE FIBONACCI ===\n")
        print(f"Términos solicitados: {self.n}\n")

        anterior = 0
        actual = 1

        if self.n == 1:
            print(anterior)
        else:
            print(anterior)
            print(actual)

            for _ in range(3, self.n + 1):
                siguiente = anterior + actual
                print(siguiente)
                anterior = actual
                actual = siguiente

        print("\nProceso completado.\n")


# ============================================================
# MAIN
# ============================================================

def main() -> None:
    try:
        numero = int(input("Ingrese la cantidad de términos: "))
        ejercicio = Ejercicio14(numero)
        ejercicio.resolver()

    except ValueError:
        print("Debe ingresar un número entero válido.")


if __name__ == "__main__":
    main()
```

### Ejercicio 15 — Calculadora con menú

**Enunciado:** Crea una calculadora que ofrezca un menú con suma, resta, multiplicación, división y salir. El usuario elige la operación y el programa repite hasta que seleccione salir.

**Objetivo:** Uso de bucles con control por condición y estructuras de selección múltiple.

```python
from dataclasses import dataclass
from abc import ABC, abstractmethod


# ============================================================
# CLASE BASE (HERENCIA + POLIMORFISMO)
# ============================================================

class EjercicioBase(ABC):
    """Clase base común para todos los ejercicios."""
    
    @abstractmethod
    def resolver(self) -> None:
        pass


# ============================================================
# EJERCICIO 15 — CALCULADORA CON MENÚ
# ============================================================

@dataclass
class Ejercicio15(EjercicioBase):
    """
    Ejercicio 15:
    Calculadora con menú que permite:
    - Suma
    - Resta
    - Multiplicación
    - División
    - Salir
    """

    _activo: bool = True   # permite controlar la ejecución del menú

    # ---------------------------------------------------------
    # ENCAPSULAMIENTO
    # ---------------------------------------------------------

    @property
    def activo(self) -> bool:
        return self._activo

    @activo.setter
    def activo(self, valor: bool) -> None:
        if isinstance(valor, bool):
            self._activo = valor
        else:
            print("El estado debe ser booleano.")

    # ---------------------------------------------------------
    # LÓGICA DEL MENÚ
    # ---------------------------------------------------------

    def resolver(self) -> None:
        print("\n=== EJERCICIO 15: CALCULADORA CON MENÚ ===\n")

        while self.activo:
            print("\nMENÚ DE OPERACIONES")
            print("1. Suma")
            print("2. Resta")
            print("3. Multiplicación")
            print("4. División")
            print("5. Salir")

            opcion = input("Seleccione una opción: ").strip()

            if opcion == "5":
                print("Saliendo del programa.")
                self.activo = False
                continue

            # Pedir números
            try:
                n1 = float(input("Ingrese el primer número: "))
                n2 = float(input("Ingrese el segundo número: "))
            except ValueError:
                print("Debe ingresar valores numéricos.")
                continue

            # Operaciones
            if opcion == "1":
                print(f"Resultado: {n1 + n2}")
            elif opcion == "2":
                print(f"Resultado: {n1 - n2}")
            elif opcion == "3":
                print(f"Resultado: {n1 * n2}")
            elif opcion == "4":
                if n2 == 0:
                    print("No es posible dividir entre cero.")
                else:
                    print(f"Resultado: {n1 / n2}")
            else:
                print("Opción no válida.")

        print("\nProceso completado.\n")


# ============================================================
# MAIN
# ============================================================

def main() -> None:
    ejercicio = Ejercicio15()
    ejercicio.resolver()


if __name__ == "__main__":
    main()
```

### Ejercicio 16 — Suma de pares hasta n

**Enunciado:** Escribe un programa que pida un número entero y calcule la suma de todos los números pares desde 1 hasta ese número (inclusive).

**Objetivo:** Acumuladores, condicionales y ciclos para suma condicional en rangos.

```python
from dataclasses import dataclass
from abc import ABC, abstractmethod


# ============================================================
# CLASE BASE (HERENCIA + POLIMORFISMO)
# ============================================================

class EjercicioBase(ABC):
    """Clase base común para todos los ejercicios."""
    
    @abstractmethod
    def resolver(self) -> None:
        pass


# ============================================================
# EJERCICIO 16 — SUMA DE PARES HASTA N
# ============================================================

@dataclass
class Ejercicio16(EjercicioBase):
    """
    Ejercicio 16:
    Calcular la suma de los números pares desde 1 hasta n (inclusive).
    """

    _n: int

    # ---------------------------------------------------------
    # ENCAPSULAMIENTO
    # ---------------------------------------------------------

    @property
    def n(self) -> int:
        return self._n

    @n.setter
    def n(self, valor: int) -> None:
        if isinstance(valor, int) and valor > 0:
            self._n = valor
        else:
            print("Debe ingresar un número entero positivo.")

    # ---------------------------------------------------------
    # LÓGICA DEL EJERCICIO
    # ---------------------------------------------------------

    def resolver(self) -> None:
        print("\n=== EJERCICIO 16: SUMA DE PARES HASTA N ===\n")
        print(f"Número ingresado: {self.n}\n")

        suma_pares: int = 0

        for numero in range(1, self.n + 1):
            if numero % 2 == 0:
                suma_pares += numero

        print(f"La suma de los números pares desde 1 hasta {self.n} es: {suma_pares}")
        print("\nProceso completado.\n")


# ============================================================
# MAIN
# ============================================================

def main() -> None:
    try:
        valor = int(input("Ingrese un número entero positivo: "))
        ejercicio = Ejercicio16(valor)
        ejercicio.resolver()

    except ValueError:
        print("Debe ingresar un número entero válido.")


if __name__ == "__main__":
    main()
```

### Ejercicio 17 — Potencia mediante multiplicación iterativa

**Enunciado:** Crea un programa que calcule base^expo sin usar el operador **. El usuario ingresa base (float) y exponente (entero no negativo).

**Objetivo:** Implementación de multiplicación repetida y uso de acumuladores en bucles.

```python
from dataclasses import dataclass
from abc import ABC, abstractmethod


# ============================================================
# CLASE BASE (HERENCIA + POLIMORFISMO)
# ============================================================

class EjercicioBase(ABC):
    """Clase base común para todos los ejercicios."""
    
    @abstractmethod
    def resolver(self) -> None:
        pass


# ============================================================
# EJERCICIO 17 — POTENCIA SIN USAR **
# ============================================================

@dataclass
class Ejercicio17(EjercicioBase):
    """
    Ejercicio 17:
    Calcular base^exponente sin usar el operador '**'.
    """

    _base: float
    _exponente: int

    # ---------------------------------------------------------
    # ENCAPSULAMIENTO
    # ---------------------------------------------------------

    @property
    def base(self) -> float:
        return self._base

    @base.setter
    def base(self, valor: float) -> None:
        if isinstance(valor, (int, float)):
            self._base = valor
        else:
            print("La base debe ser un número válido.")

    @property
    def exponente(self) -> int:
        return self._exponente

    @exponente.setter
    def exponente(self, valor: int) -> None:
        if isinstance(valor, int) and valor >= 0:
            self._exponente = valor
        else:
            print("El exponente debe ser un número entero no negativo.")

    # ---------------------------------------------------------
    # LÓGICA DEL EJERCICIO
    # ---------------------------------------------------------

    def resolver(self) -> None:
        print("\n=== EJERCICIO 17: POTENCIA SIN USAR ** ===\n")
        print(f"Base: {self.base}")
        print(f"Exponente: {self.exponente}\n")

        resultado: float = 1.0

        for _ in range(self.exponente):
            resultado *= self.base

        print(f"Resultado de {self.base}^{self.exponente}: {resultado}")
        print("\nProceso completado.\n")


# ============================================================
# MAIN
# ============================================================

def main() -> None:
    try:
        base = float(input("Ingrese la base: "))
        exponente = int(input("Ingrese el exponente (entero no negativo): "))

        ejercicio = Ejercicio17(base, exponente)
        ejercicio.resolver()

    except ValueError:
        print("Debe ingresar valores numéricos válidos.")


if __name__ == "__main__":
    main()
```

```python
from dataclasses import dataclass
from abc import ABC, abstractmethod


# ============================================================
# CLASE BASE (HERENCIA + POLIMORFISMO)
# ============================================================

class EjercicioBase(ABC):
    @abstractmethod
    def resolver(self) -> None:
        pass


# ============================================================
# EJERCICIO INDUSTRIA 1 — CONTROL DE INVENTARIO
# ============================================================

@dataclass
class EjercicioIndustria1(EjercicioBase):
    """
    Control de inventario: calcular inventario final y unidades faltantes.
    """

    _inventario_inicial: int
    _vendidas: int

    # ---------------------------------------------------------
    # ENCAPSULAMIENTO
    # ---------------------------------------------------------
    @property
    def inventario_inicial(self) -> int:
        return self._inventario_inicial

    @inventario_inicial.setter
    def inventario_inicial(self, valor: int) -> None:
        if isinstance(valor, int) and valor >= 0:
            self._inventario_inicial = valor
        else:
            print("El inventario inicial debe ser un entero no negativo.")

    @property
    def vendidas(self) -> int:
        return self._vendidas

    @vendidas.setter
    def vendidas(self, valor: int) -> None:
        if isinstance(valor, int) and valor >= 0:
            self._vendidas = valor
        else:
            print("Las unidades vendidas deben ser un entero no negativo.")

    # ---------------------------------------------------------
    # LÓGICA DEL EJERCICIO
    # ---------------------------------------------------------
    def resolver(self) -> None:
        print("\n=== CONTROL DE INVENTARIO ===\n")
        print(f"Inventario inicial: {self.inventario_inicial}")
        print(f"Unidades vendidas: {self.vendidas}\n")

        restante: int = self.inventario_inicial - self.vendidas

        if restante >= 0:
            print(f"Inventario final: {restante}")
        else:
            faltantes: int = abs(restante)
            print("Inventario final: 0")
            print(f"Faltaron {faltantes} unidades para cubrir toda la demanda.")

        print("\nProceso completado.\n")


# ============================================================
# MAIN
# ============================================================

def main() -> None:
    try:
        inventario = int(input("Ingrese el inventario inicial: "))
        vendidas = int(input("Ingrese la cantidad vendida: "))

        ejercicio = EjercicioIndustria1(inventario, vendidas)
        ejercicio.resolver()

    except ValueError:
        print("Debe ingresar valores numéricos enteros.")


if __name__ == "__main__":
    main()
```

### Ejercicio 27 — Cálculo de desperdicio**

**Enunciado:**
Una máquina consume materia prima y genera desperdicio. El sistema recibe la cantidad de materia prima inicial y final, y debe calcular el porcentaje de desperdicio.



```python
from dataclasses import dataclass
from abc import ABC, abstractmethod


class EjercicioBase(ABC):
    @abstractmethod
    def resolver(self) -> None:
        pass


@dataclass
class EjercicioIndustria10(EjercicioBase):
    _inicial: float
    _final: float

    @property
    def inicial(self) -> float:
        return self._inicial

    @inicial.setter
    def inicial(self, valor: float) -> None:
        if isinstance(valor, (int, float)) and valor >= 0:
            self._inicial = valor

    @property
    def final(self) -> float:
        return self._final

    @final.setter
    def final(self, valor: float) -> None:
        if isinstance(valor, (int, float)) and valor >= 0:
            self._final = valor

    def resolver(self) -> None:
        print("\n=== CÁLCULO DE DESPERDICIO ===\n")
        desperdicio = self.inicial - self.final
        porcentaje = (desperdicio / self.inicial) * 100 if self.inicial > 0 else 0

        print(f"Materia prima inicial: {self.inicial}")
        print(f"Materia prima final: {self.final}")
        print(f"Desperdicio total: {desperdicio}")
        print(f"Porcentaje de desperdicio: {porcentaje:.2f}%")


def main() -> None:
    ini = float(input("Ingrese cantidad inicial: "))
    fin = float(input("Ingrese cantidad final: "))
    ejercicio = EjercicioIndustria10(ini, fin)
    ejercicio.resolver()


if __name__ == "__main__":
    main()
```
