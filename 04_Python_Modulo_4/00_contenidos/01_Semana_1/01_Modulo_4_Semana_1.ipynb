# 1. ¿Qué es la persistencia realmente? (Concepto clave)

Un programa tiene dos memorias:

| Tipo                | Característica                 | Ejemplo                  |
| ------------------- | ------------------------------ | ------------------------ |
| Memoria volátil     | Se borra al cerrar el programa | variables                |
| Memoria persistente | Sobrevive al programa          | archivos, bases de datos |

Ejemplo mental:

```
Programa de notas:

Sin archivo:
Escribes -> cierras -> desaparece

Con archivo:
Escribes -> cierras -> vuelve a aparecer
```

Entonces:

> Persistir = transferir información de RAM a almacenamiento permanente

## Analogía pedagógica

La RAM es como la mente
El archivo es como un cuaderno

Si no escribes en el cuaderno → olvidas

# 3. Bloque 2 — Infraestructura reutilizable de archivos

## util: file_manager.py

Aquí aplicamos **DRY + SRP + seguridad**


````python
from pathlib import Path            # Permite manejar rutas de archivos de forma segura y multiplataforma
from typing import List             # Permite indicar tipos de datos (en este caso lista de strings)

class FileManager:                  # Clase encargada de gestionar archivos de texto
    """Gestiona operaciones seguras sobre archivos de texto."""

    def __init__(self, file_path: str) -> None:   # Constructor: recibe la ruta del archivo
        self._path = Path(file_path)              # Convierte el string en un objeto Path
        self._ensure_file()                       # Garantiza que el archivo exista antes de usarlo

    def _ensure_file(self) -> None:               # Método interno (privado) para asegurar existencia
        """Garantiza que el archivo exista."""
        self._path.parent.mkdir(parents=True, exist_ok=True)  # Crea carpetas si no existen
        self._path.touch(exist_ok=True)                        # Crea el archivo si no existe

    def read_lines(self) -> List[str]:           # Lee el archivo y devuelve una lista de líneas
        try:                                     # Bloque de control de errores
            with self._path.open("r", encoding="utf-8") as file:  # Abre archivo en modo lectura
                return [line.strip() for line in file.readlines()] # Lee todas las líneas y elimina saltos
        except OSError as error:                 # Si ocurre un error del sistema de archivos
            raise RuntimeError("Error leyendo archivo") from error # Lanza error más claro para la app

    def append_line(self, text: str) -> None:    # Agrega una nueva línea al archivo
        if "\n" in text:                         # Valida que no tenga saltos de línea internos
            raise ValueError("El texto no debe contener saltos de línea") # Evita corrupción del archivo

        try:                                     # Manejo de errores de escritura
            with self._path.open("a", encoding="utf-8") as file:  # Abre archivo en modo añadir (append)
                file.write(text + "\n")          # Escribe el texto y agrega salto de línea manualmente
        except OSError as error:                 # Si falla escritura en disco
            raise RuntimeError("Error escribiendo archivo") from error # Error controlado para el sistema
````

Aquí el estudiante aprende:

* Encapsulación
* Tipado
* Manejo de errores
* Reutilización
* Seguridad

# 3. Bloque 3 — Modelo de dominio

## models/model_task.py


````python
from dataclasses import dataclass          # Permite crear clases de datos automáticamente (menos código boilerplate)

@dataclass(frozen=True)                    # Genera constructor, equals y repr; frozen=True lo hace inmutable
class Task:                                # Representa una tarea del sistema
    title: str                             # Nombre o descripción de la tarea
    completed: bool = False                # Estado de la tarea (por defecto no completada)

    def serialize(self) -> str:            # Convierte el objeto a texto para guardarlo en archivo
        return f"{self.title}|{int(self.completed)}"  # Guarda como: titulo|0 o titulo|1

    @staticmethod                          # Método de clase sin necesidad de instancia previa
    def deserialize(data: str) -> "Task":  # Convierte una línea de texto nuevamente en objeto Task
        title, completed = data.split("|") # Separa los datos usando el separador |
        return Task(title=title, completed=bool(int(completed)))  # Reconstruye el objeto convirtiendo 0/1 a booleano
````

Conceptos enseñados:

* Inmutabilidad
* Serialización manual
* Representación persistente

# 7. Bloque 4 — Patrón Repository

Separar negocio de almacenamiento

## repository/repository_task.py


````python
from typing import List                     # Permite tipar listas para mayor claridad y control
from models.model_task import Task                # Importa el modelo de datos Task
from utils.file_manager import FileManager # Importa la capa encargada de manejar archivos

class TaskRepository:                       # Capa repositorio: comunica el sistema con el almacenamiento
    """Repositorio basado en archivo plano."""

    def __init__(self, file_path: str) -> None:  # Constructor que recibe la ruta del archivo
        self._file = FileManager(file_path)      # Inicializa el gestor de archivos

    def get_all(self) -> List[Task]:        # Obtiene todas las tareas almacenadas
        lines = self._file.read_lines()     # Lee todas las líneas del archivo
        return [Task.deserialize(line) for line in lines if line]  # Convierte cada línea en objeto Task

    def save(self, task: Task) -> None:     # Guarda una nueva tarea en el archivo
        self._file.append_line(task.serialize()) # Convierte el objeto a texto y lo agrega al archivo
````

---

### Concepto clave enseñado

El programa **no sabe que existe un archivo**

→ solo sabe que existe un repositorio

Esto es arquitectura profesional.

# 3. Bloque 5 — Validación y capa de servicio

## utils/validator.py



````python
def validate_title(title: str) -> str:      # Función que valida el texto del título de la tarea
    title = title.strip()                   # Elimina espacios al inicio y al final

    if not title:                           # Verifica que no quede vacío
        raise ValueError("La tarea no puede estar vacía")  # Error si el texto está vacío

    if len(title) > 60:                     # Verifica longitud máxima permitida
        raise ValueError("La tarea es demasiado larga")    # Error si supera 60 caracteres

    if "|" in title:                        # Evita el uso del separador del archivo
        raise ValueError("Caracter inválido")              # Error si contiene '|'

    return title                            # Retorna el título limpio y válido
````

## services/service_task.py


````python
from repository.repository_task import TaskRepository  # Importa la capa de acceso a datos
from models.model_task import Task                     # Importa el modelo de tarea
from utils.validator import validate_title             # Importa la función de validación

class TaskService:                                     # Capa de negocio: coordina validaciones y repositorio

    def __init__(self, repository: TaskRepository) -> None:  # Recibe el repositorio por inyección de dependencia
        self._repository = repository                        # Guarda referencia al repositorio

    def create_task(self, title: str) -> None:               # Caso de uso: crear una tarea
        clean_title = validate_title(title)                  # Valida el título antes de procesar
        task = Task(clean_title)                             # Crea el objeto de dominio
        self._repository.save(task)                          # Persiste la tarea

    def list_tasks(self) -> list[Task]:                      # Caso de uso: listar tareas
        return self._repository.get_all()                    # Obtiene todas las tareas almacenadas
````

---

### Conceptos enseñados

* Arquitectura en capas
* Regla de negocio ≠ persistencia
* Validación previa a guardar (Fail Fast)

# 3. Bloque 6 — Programa principal

## main.py


````python
from services.service_task import TaskService        # Importa la capa de lógica de negocio
from repository.repository_task import TaskRepository # Importa la capa de acceso a datos

def main() -> None:                                  # Función principal del programa
    service = TaskService(TaskRepository("data/tasks.txt"))  # Inicializa servicio con repositorio y archivo

    while True:                                      # Bucle infinito para menú interactivo
        print("\n1. Agregar tarea")                  # Opción 1 del menú
        print("2. Listar tareas")                    # Opción 2 del menú
        print("3. Salir")                            # Opción 3 del menú

        option = input("Seleccione: ")               # Lee la opción elegida por el usuario

        if option == "1":                            # Si elige agregar tarea
            title = input("Tarea: ")                 # Solicita el nombre de la tarea
            try:                                     # Manejo de errores controlado
                service.create_task(title)           # Ejecuta el caso de uso crear tarea
                print("Guardado correctamente")      # Mensaje de éxito
            except Exception as e:                   # Captura cualquier error validado previamente
                print("Error:", e)                   # Muestra mensaje limpio al usuario

        elif option == "2":                          # Si elige listar tareas
            tasks = service.list_tasks()             # Obtiene las tareas del sistema
            for i, task in enumerate(tasks, 1):      # Recorre las tareas con índice iniciando en 1
                print(f"{i}. {task.title}")          # Muestra cada tarea en pantalla

        elif option == "3":                          # Si elige salir
            break                                    # Termina el programa


if __name__ == "__main__":                           # Punto de entrada del script
    main()                                           # Ejecuta la aplicación
````

## utils/file_manager.py MODIFICADO

````python
import tempfile      # Permite crear archivos temporales seguros
import os            # Permite operaciones del sistema como reemplazar archivos

def overwrite_all(self, lines: List[str]) -> None:     # Sobrescribe el archivo completo de forma segura
    try:
        with tempfile.NamedTemporaryFile(              # Crea archivo temporal en la misma carpeta
            mode="w",                                  # Modo escritura
            encoding="utf-8",                          # Codificación estándar
            delete=False,                              # No se borra al cerrar (lo usaremos luego)
            dir=self._path.parent                      # Mismo directorio para permitir reemplazo atómico
        ) as tmp:
            for line in lines:                         # Recorre todas las líneas a guardar
                tmp.write(line + "\n")                 # Escribe cada línea en el temporal

        os.replace(tmp.name, self._path)               # Reemplaza el archivo original por el temporal (operación atómica)

    except OSError as error:                           # Captura errores del sistema de archivos
        raise RuntimeError("Error sobrescribiendo archivo") from error  # Lanza error controlado
````

## repository/repository_task.py MODIFICADO

````python
def overwrite_all(self, tasks: List[Task]) -> None:     # Recibe todas las tareas a guardar
    serialized = [task.serialize() for task in tasks]   # Convierte cada objeto Task a texto almacenable
    self._file.overwrite_all(serialized)                # Delegá la escritura segura al FileManager
````

## utils/validator.py MODIFICADO

````python
def validate_index(index: int, size: int) -> int:
    if size == 0:
        raise ValueError("No hay tareas registradas")

    if index < 1:
        raise ValueError("El número debe ser mayor que cero")

    if index > size:
        raise ValueError("Número fuera de rango")

    return index - 1
````

## services/service_task.py MODIFICADO

````python
from typing import List                          # Permite tipar listas

def complete_task(self, index: int) -> None:     # Marca una tarea como completada
    tasks = self._repository.get_all()           # Obtiene todas las tareas almacenadas

    validated_index = validate_index(index, len(tasks))  # Valida que el número ingresado sea correcto

    updated_task = tasks[validated_index].mark_completed()  # Crea nueva instancia marcada como completada
    tasks[validated_index] = updated_task                   # Reemplaza la tarea en la lista

    self._repository.overwrite_all(tasks)         # Guarda nuevamente todas las tareas sin corromper archivo
````
