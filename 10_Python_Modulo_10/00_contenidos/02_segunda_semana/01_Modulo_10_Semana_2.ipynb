# Prompt Maestro — Sistema de Testing Frontend con Vitest + React Testing Library

---



## PROMPT

Actúa como un **Frontend Test Engineer Senior especializado en React, Testing Library y pruebas de comportamiento (behavior-driven testing)**.

Debes diseñar e implementar un **sistema completo de pruebas automatizadas para una aplicación React moderna** usando:

* Vitest
* React Testing Library
* user-event

El objetivo NO es solo comprobar funciones internas, sino validar el comportamiento observable desde la perspectiva del usuario.

El resultado debe ser didáctico pero alineado con prácticas reales de la industria frontend.

---

## 1. Principio rector

Todas las pruebas deben seguir la filosofía:

> “Mientras más se parezca al uso real del usuario, más confiable es el test.”

Evita:

* probar estado interno
* probar funciones privadas
* usar querySelector
* usar testId innecesariamente

Prioriza:

* roles
* texto visible
* accesibilidad
* interacción realista

---

## 2. Componentes a probar

El sistema contiene:

### TaskForm

Formulario para crear tareas:

* input texto
* botón enviar
* validación
* limpieza del campo tras submit
* callback onCreateTask

---

### TaskList

Lista de tareas:

* renderiza múltiples tareas
* estado vacío
* marcado como completado
* eliminación
* renderización condicional

Si falta comportamiento, propón uno razonable coherente con apps reales.

---

## 3. Tipos de pruebas requeridas

Debes diseñar:

### 3.1 Unit behavior tests

Validan comportamiento del componente aislado

Ejemplos:

* render inicial
* contenido visible
* accesibilidad
* props obligatorias

---

### 3.2 Interaction tests

Simulan acciones del usuario

Usar obligatoriamente:

```
userEvent
```

Simular:

* escribir texto
* click
* submit
* toggle
* borrar

---

### 3.3 Functional UI tests

Validan cambios de interfaz tras interacción

Ejemplos:

* aparece error si input vacío
* tarea agregada aparece en lista
* botón deshabilitado
* estado completado cambia estilo

---

## 4. Mocking requerido

Debes usar mocks para:

* callbacks (onCreateTask, onDelete, onToggle)
* funciones async simuladas
* datos de servidor simulados

Explica cuándo usar:

* vi.fn()
* vi.mock()

---

## 5. Organización del proyecto de pruebas

Debes crear estructura:

```
src/
components/
tests/
    setup/
    unit/
    interaction/
    functional/
    fixtures/
```

Explica el propósito de cada carpeta.

---

## 6. Buenas prácticas obligatorias

El sistema debe aplicar:

* patrón Arrange-Act-Assert
* screen en lugar de container
* queries por rol antes que por texto
* await en interacciones
* tests legibles (nombre como historia)
* independencia entre pruebas
* evitar over-testing

---

## 7. Accesibilidad

Los tests deben validar accesibilidad:

* roles correctos
* labels asociados
* botones accesibles
* navegación por teclado simulada

---

## 8. Casos obligatorios a cubrir

Para TaskForm:

* input vacío
* texto válido
* texto largo
* espacios en blanco
* múltiples submits
* limpieza del campo
* submit por Enter

Para TaskList:

* lista vacía
* múltiples tareas
* toggle completado
* eliminación
* re-render

---

## 9. Configuración del entorno

Debes crear:

* configuración Vitest
* setup global
* jsdom
* testing library matchers

Explicar cada archivo creado.

## 10. Uso de IA como copiloto

Incluye instrucciones para que un estudiante use IA para:

* generar casos faltantes

* simplificar tests largos
* refactorizar duplicación

Debes mostrar ejemplos de prompts útiles.

## 11. Resultado esperado

Debes entregar:

1. estructura completa de pruebas
2. configuración Vitest funcional

4. mocks correctos
5. interacción realista
