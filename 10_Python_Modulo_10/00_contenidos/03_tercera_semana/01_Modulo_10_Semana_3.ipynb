PROMPT

Actúa como **Senior QA Automation Engineer + SDET + Software Architect** especializado en testing de aplicaciones web distribuidas modernas.

Tu tarea es **diseñar e implementar un suite completo profesional de pruebas end-to-end usando Cypress** para un sistema web tipo **Universidad Digital** compuesto por:

Frontend: React SPA
Backend: API REST
Base de datos persistente

El objetivo es **crear pruebas reales listas para ejecutarse en un proyecto profesional**.

---

## OBJETIVO DEL SUITE

Validar el flujo completo del usuario:

login → navegación → creación de tarea → confirmación → persistencia → visualización

El suite debe detectar defectos en:

* comunicación frontend/backend
* serialización de datos
* validaciones del backend
* renderización del frontend
* estados inconsistentes
* errores HTTP
* latencias
* errores silenciosos

---

## REQUISITOS DE IMPLEMENTACIÓN

### 1) Estructura profesional del proyecto

Debes crear:

* estructura de carpetas Cypress escalable
* custom commands
* fixtures
* page objects (si aplica)
* helpers reutilizables
* configuración por ambientes
* baseUrl configurable
* manejo de credenciales por env

No generar archivos monolíticos.

---

### 2) Configuración técnica obligatoria

Configurar correctamente:

* cypress.config
* variables de entorno
* intercepts globales
* manejo de sesiones
* limpieza de estado entre pruebas
* retry strategy
* timeouts realistas

---

### 3) Escenarios a automatizar

#### Autenticación

* login exitoso
* login inválido
* credenciales vacías
* token expirado
* backend caído durante login

#### Navegación

* acceso a dashboard autenticado
* bloqueo sin autenticación

#### Creación de tarea

* crear tarea válida
* crear tarea con datos inválidos
* backend retorna error 500
* backend lento
* pérdida de conexión

#### Confirmación

* validación visual
* validación por API
* validación de persistencia

#### Visualización

* tarea aparece en lista
* datos correctos
* refresco mantiene estado

---

### 4) Validaciones obligatorias

Cada test debe validar:

UI + API + datos

No solo elementos HTML.

Debe validar:

* status codes
* body JSON
* tiempo de respuesta
* consistencia de datos renderizados

---

### 5) Uso avanzado de Cypress

Debes usar correctamente:

* cy.intercept
* aliases
* waits inteligentes (no waits fijos)
* comandos personalizados
* fixtures dinámicos
* beforeEach / afterEach
* manejo de sesiones

NO usar:

cy.wait(5000)

---

### 6) Detección de fallos reales

El suite debe incluir pruebas que simulen:

* backend caído
* respuesta corrupta
* timeout
* validaciones inconsistentes
* respuesta parcial

---

### 7) Calidad del código

El código generado debe:

* estar modularizado
* ser legible
* tener nombres semánticos
* evitar duplicación
* permitir escalar a cientos de pruebas
