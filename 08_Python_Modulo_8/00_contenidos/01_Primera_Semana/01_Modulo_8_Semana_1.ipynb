### **1.2 Fundamentos tĂ©cnicos**

#### **Declarativo vs. Imperativo**

* En un enfoque **imperativo**, el programador le indica al navegador cada paso para actualizar la UI.
* En un enfoque **declarativo (React)**, el programador describe el *estado final deseado*, y React actualiza la interfaz automĂˇticamente cuando el estado cambia.

**Ejemplo comparativo:**

| Estilo               | Ejemplo                                                       | Desventaja                                |
| -------------------- | ------------------------------------------------------------- | ----------------------------------------- |
| Imperativo (JS puro) | `document.getElementById('app').innerHTML = '<h1>Hola</h1>';` | DifĂ­cil de mantener y propenso a errores. |
| Declarativo (React)  | `<h1>Hola</h1>` dentro de un componente                       | MĂˇs legible, mantenible y predecible.     |

#### **Virtual DOM**

**Importante:** El DOM (Modelo de Objetos de Documento) es la representaciĂłn en forma de Ăˇrbol de una pĂˇgina web que el navegador crea para interactuar con sus elementos. 

El **Virtual DOM** es una **representaciĂłn en memoria del DOM real**.
Cuando ocurre un cambio, React **compara (diff)** el Virtual DOM con el DOM anterior y **actualiza solo los elementos modificados**, optimizando el rendimiento.

**Ventajas tĂ©cnicas:**

* Reduce operaciones costosas en el navegador.
* Mejora la eficiencia (ISO/IEC 25010: eficiencia de ejecuciĂłn).
* Aumenta la responsividad y estabilidad visual.

*Ejemplo conceptual:*

```
Estado anterior â†’ <h1>Hola</h1>
Estado nuevo    â†’ <h1>Hola Mundo</h1>
React actualiza solo la palabra â€śMundoâ€ť en el DOM real.
```

#### **Unidireccionalidad de datos**

En React, los datos **fluyen en una sola direcciĂłn**:

Esto asegura:
