## **1. Introducción: ¿Qué es React y por qué se usa en el frontend profesional?**

### **1.1 ¿Qué es React?**

React es una **librería JavaScript** creada por Facebook (Meta) en 2013, diseñada para construir **interfaces de usuario interactivas y eficientes**.

Se diferencia de los frameworks tradicionales (como Angular o Vue) porque:

* Está enfocada únicamente en la **capa de la vista (View)** dentro del modelo MVC.
* Permite integrar otras librerías libremente (es **de naturaleza modular y flexible**).
* Utiliza una **arquitectura declarativa**, lo que significa que el desarrollador define *qué quiere que se vea*, y React decide *cómo renderizarlo de forma eficiente*.

**Definición formal (ISO/IEC 25010 - mantenibilidad):**

> React promueve la mantenibilidad del software al organizar la interfaz en componentes reutilizables y con responsabilidades bien definidas.

---


### **1.2 Fundamentos técnicos**

#### **Declarativo vs. Imperativo**

* En un enfoque **imperativo**, el programador le indica al navegador cada paso para actualizar la UI.
* En un enfoque **declarativo (React)**, el programador describe el *estado final deseado*, y React actualiza la interfaz automáticamente cuando el estado cambia.

**Ejemplo comparativo:**

| Estilo               | Ejemplo                                                       | Desventaja                                |
| -------------------- | ------------------------------------------------------------- | ----------------------------------------- |
| Imperativo (JS puro) | `document.getElementById('app').innerHTML = '<h1>Hola</h1>';` | Difícil de mantener y propenso a errores. |
| Declarativo (React)  | `<h1>Hola</h1>` dentro de un componente                       | Más legible, mantenible y predecible.     |



---

#### **Virtual DOM**

**Importante:** El DOM (Modelo de Objetos de Documento) es la representación en forma de árbol de una página web que el navegador crea para interactuar con sus elementos. 

El **Virtual DOM** es una **representación en memoria del DOM real**.
Cuando ocurre un cambio, React **compara (diff)** el Virtual DOM con el DOM anterior y **actualiza solo los elementos modificados**, optimizando el rendimiento.

**Ventajas técnicas:**

* Reduce operaciones costosas en el navegador.
* Mejora la eficiencia (ISO/IEC 25010: eficiencia de ejecución).
* Aumenta la responsividad y estabilidad visual.

*Ejemplo conceptual:*

```
Estado anterior → <h1>Hola</h1>
Estado nuevo    → <h1>Hola Mundo</h1>
React actualiza solo la palabra “Mundo” en el DOM real.
```

---

#### **Unidireccionalidad de datos**

En React, los datos **fluyen en una sola dirección**:
del **componente padre → al hijo**, a través de las **props**.

Esto asegura:

* Control predecible del estado.
* Facilidad de depuración.
* Evita efectos secundarios inesperados.

**Norma internacional aplicada:**

> Cumple con el principio de *cohesión alta y acoplamiento bajo* (ISO/IEC 25010, mantenibilidad y modularidad).

---


### **1.3. Uso de `useState`, `props` y manejo de eventos en React**

En React, el **ciclo interactivo entre interfaz, datos y usuario** se basa en tres pilares fundamentales:

1. **`useState` →** permite manejar datos dinámicos (estado).
2. **`props` →** permite comunicar información entre componentes.
3. **Manejo de eventos →** permite responder a las acciones del usuario (clics, envíos, cambios, etc.).

Estos tres elementos son la base del **paradigma declarativo** de React, que facilita la construcción de **interfaces reactivas, reutilizables y seguras**.

---

#### **1.3.1. Teoría de `useState`**

### Definición

`useState` es un **Hook** de React que permite a los componentes funcionales **recordar información a lo largo del tiempo** y **actualizar la interfaz** cuando esos datos cambian.

En otras palabras, es la forma en la que un componente “recuerda” su propio estado interno sin tener que volver a renderizar toda la aplicación.

---

#### **Concepto clave: Estado en React**

El **estado** es una estructura de datos que **representa el valor actual** de algo que puede cambiar con el tiempo:

* Datos de formularios
* Listas dinámicas
* Contadores
* Estados de sesión, temas, etc.

React actualiza automáticamente la interfaz cada vez que el estado cambia, manteniendo la **sincronía entre la vista y los datos**.

---

#### Sintaxis general

```jsx
const [estado, setEstado] = useState(valorInicial);
```

* `estado`: variable que contiene el valor actual.
* `setEstado`: función que actualiza el valor y **provoca un nuevo renderizado del componente**.
* `useState(valorInicial)`: inicializa el estado con un valor por defecto.

---

#### Principios teóricos asociados

| Principio                          | Aplicación                                                                                                  |
| ---------------------------------- | ----------------------------------------------------------------------------------------------------------- |
| **Inmutabilidad**                  | No se modifica el estado directamente; se genera una nueva copia para evitar errores y efectos colaterales. |
| **Unidireccionalidad de datos**    | Los datos fluyen del componente padre al hijo; los hijos no deben alterar el estado del padre directamente. |
| **Reactividad declarativa**        | React vuelve a renderizar automáticamente cuando el estado cambia, sin manipular el DOM manualmente.        |
| **Mantenibilidad (ISO/IEC 25010)** | Permite componentes predecibles y fáciles de probar.                                                        |

---

#### Ejemplo conceptual 

Supongamos un sistema donde un usuario marca tareas como completadas.
Cada tarea tiene un **estado interno** (`completada = true/false`).

Al usar `useState`, el componente:

* Almacena si la tarea está completada o no.
* Actualiza la interfaz visual automáticamente cuando el usuario interactúa.

---



````python
import React, { useState } from 'react';

function Contador() {
  // Declaración del estado: contador inicia en 0
  const [contador, setContador] = useState(0);

  // Función para aumentar el contador
  const incrementar = () => setContador(contador + 1);

  return (
    <div style={{ textAlign: 'center', marginTop: '20px' }}>
      <h2>Contador: {contador}</h2>
      <button onClick={incrementar}>Aumentar</button>
    </div>
  );
}

export default Contador;
````

### **1.3.2. Teoría de `props` (propiedades)**

#### Definición

Los **props** (abreviatura de *properties*) son los **parámetros de entrada** que recibe un componente React.
Permiten **transferir datos desde un componente padre hacia uno hijo**, haciendo que los componentes sean **reutilizables y dinámicos**.

---

#### Fundamento conceptual

React sigue el principio de **composición**: los componentes se construyen combinando otros componentes más pequeños.
Los `props` permiten pasar información entre ellos, similar a cómo las funciones reciben argumentos.

Por ejemplo:

* Un componente padre (`App`) envía un valor o función a un componente hijo (`TaskItem`).
* El hijo recibe esos datos mediante `props` y los usa para mostrar o ejecutar acciones.

---

#### Características teóricas de los `props`

| Característica       | Descripción                                                                                   |
| -------------------- | --------------------------------------------------------------------------------------------- |
| **Inmutables**       | Un componente no puede modificar directamente los `props` que recibe.                         |
| **Unidireccionales** | Los datos fluyen en una sola dirección: de padre → hijo.                                      |
| **Deterministas**    | Un mismo conjunto de `props` siempre produce el mismo resultado visual (principio de pureza). |
| **Reutilización**    | Facilitan crear componentes genéricos que se pueden usar en distintos contextos.              |

---

#### Relación con normas internacionales

* **ISO/IEC 25010 – Calidad del software:** los `props` promueven **mantenibilidad** y **reusabilidad**.
* **SOLID (Principio de Responsabilidad Única - SRP):** cada componente debe recibir solo los `props` que necesita.
* **DRY (Don’t Repeat Yourself):** los `props` evitan duplicación de código en componentes repetidos.

---



#### **Ejemplo de Props: Saludo personalizado**

### Componente hijo: `Saludo.jsx`


````python
import React from 'react';

/**
 * Componente que recibe un prop "nombre" y lo muestra
 */
function Saludo({ nombre }) {
  return <h2>Hola, {nombre} 👋</h2>;
}

export default Saludo;
````

### Componente padre: `App.jsx`



````python
import React from 'react';
import Saludo from './components/Saludo';

function App() {
  return (
    <div style={{ textAlign: 'center', marginTop: '20px' }}>
      <h1>Mi App React</h1>
      <Saludo nombre="Luis" />
      <Saludo nombre="María" />
      <Saludo nombre="Pedro" />
    </div>
  );
}

export default App;
````

#### **1.3.3. Teoría del manejo de eventos**

#### Definición

El manejo de eventos en React es el mecanismo que permite **responder a las interacciones del usuario**, como:

* Clicks
* Cambios en un formulario
* Presionar teclas
* Envío de formularios, entre otros.

React usa un sistema de **eventos sintéticos** (Synthetic Events), que **normaliza los eventos del navegador** para que funcionen igual en todos los entornos.

---

#### **Concepto técnico: SyntheticEvent**

> `SyntheticEvent` es una abstracción del evento del navegador que garantiza compatibilidad y rendimiento.
> Permite escribir un mismo código de interacción que funciona de forma segura y uniforme en todos los navegadores.

---

####  **Principios teóricos de los eventos en React**

| Principio                                            | Descripción                                                                                                                            |
| ---------------------------------------------------- | -------------------------------------------------------------------------------------------------------------------------------------- |
| **Declaratividad**                                   | En lugar de modificar el DOM directamente, React define cómo debería reaccionar la interfaz.                                           |
| **Inmutabilidad**                                    | Los eventos no alteran el estado directamente; se usan funciones que actualizan el estado con `setState` o `useState`.                 |
| **Unidireccionalidad**                               | Los eventos pueden invocar funciones del componente padre mediante `props`, pero los datos fluyen siempre en una dirección controlada. |
| **Seguridad (OWASP A05: Security Misconfiguration)** | No se deben incluir funciones inline con contenido sensible ni manipulación directa del DOM que exponga vulnerabilidades XSS.          |

---

#### **1.3.4. Integración teórica de los tres conceptos**

#### **Relación entre ellos**

| Concepto     | Función principal                               | Interacción con los demás                                        |
| ------------ | ----------------------------------------------- | ---------------------------------------------------------------- |
| **useState** | Mantiene datos dinámicos dentro del componente. | Cambia el valor del estado cuando ocurre un evento.              |
| **props**    | Permite pasar datos entre componentes.          | Un padre puede pasar funciones o valores al hijo como `props`.   |
| **Eventos**  | Capturan las acciones del usuario.              | Ejecutan funciones que actualizan el estado mediante `useState`. |


#### **Resumen de cada definición**




| Característica          | **useState**                                                                                 | **props**                                                                                    | **Manejo de eventos**                                                                                                     |
| ----------------------- | -------------------------------------------------------------------------------------------- | -------------------------------------------------------------------------------------------- | ------------------------------------------------------------------------------------------------------------------------- |
| **Definición**          | Hook que permite **almacenar y gestionar el estado interno** de un componente.               | Mecanismo para **pasar datos de un componente padre a un hijo**.                             | Sistema que permite **responder a interacciones del usuario** (clics, cambios, envíos de formularios) mediante funciones. |
| **Origen**              | Declarado dentro del **componente que lo usa**.                                              | Se recibe desde el **componente padre**.                                                     | Definido dentro del **componente**, se asigna a un evento del JSX (`onClick`, `onChange`, `onSubmit`).                    |
| **Mutabilidad**         | **Mutable**: se actualiza con la función `setEstado`.                                        | **Inmutable**: el hijo no puede modificar los props recibidos.                               | La función manejadora puede **modificar el estado interno** o ejecutar lógica, pero no cambia directamente los props.     |
| **Alcance**             | Solo afecta al **componente que lo declara**.                                                | Afecta al **componente hijo** que recibe los datos.                                          | Puede afectar **el componente local** (con `useState`) o ejecutar **callbacks en el padre** mediante props.               |
| **Función principal**   | Mantener información **dinámica** que puede cambiar con interacciones del usuario o eventos. | Permitir **comunicación entre componentes** y reutilización de componentes.                  | Permitir **interacción y reacción a acciones del usuario**, actualizando estado, ejecutando lógica o llamadas externas.   |
| **Actualización de UI** | Cambiar el estado con `setEstado` provoca **re-render del componente**.                      | Cambiar el valor de los props en el padre provoca **re-render del hijo**.                    | Ejecutar la función del evento puede **cambiar el estado**, lo que provoca un **re-render** de los componentes afectados. |
| **Uso típico**          | Contadores, formularios, visibilidad de elementos, listas dinámicas, toggles.                | Textos personalizados, datos de listas, funciones de callback, configuración de componentes. | Botones, formularios, inputs, toggles, eventos del mouse y teclado, menús dinámicos.                                      |
| **Ejemplo breve**       | `const [contador, setContador] = useState(0)`                                                | `<Saludo nombre="Luis" />`                                                                   | `<button onClick={incrementar}>Aumentar</button>`                                                                         |

### **2. ¿Qué es Vite?**

* Es un entorno de desarrollo moderno y liviano para aplicaciones frontend

* Es mucho más rápido que Webpack o CRA (Create React App), ya que usa **ES Modules nativos** y un **servidor de desarrollo ultrarrápido**.


### **Características principales de Vite**

| Característica                             | Descripción                                                                                                           |
| ------------------------------------------ | --------------------------------------------------------------------------------------------------------------------- |
| **Servidor de desarrollo ultrarrápido**    | Vite inicia un servidor local que recarga los cambios casi al instante (Hot Module Replacement, HMR).                 |
| **Bundler optimizado para producción**     | Compila y empaqueta tu aplicación usando **Rollup** para generar archivos optimizados.                                |
| **Soporte nativo a ES Modules**            | Aprovecha los módulos de JavaScript modernos, eliminando compilaciones pesadas innecesarias durante el desarrollo.    |
| **Compatibilidad con frameworks modernos** | Funciona con **React, Vue, Svelte, Preact, Solid** y más.                                                             |
| **Plugins y extensibilidad**               | Tiene un ecosistema de plugins que permite agregar funcionalidades como TailwindCSS, TypeScript, ESLint, entre otros. |



**Ventajas clave:**

* Inicio instantáneo del servidor (sin compilación completa).
* Recarga en tiempo real (*Hot Module Replacement - HMR*).
* Integración fácil con React, Tailwind y TypeScript.
* Configuración más limpia, adaptable a normas de seguridad (sin dependencias obsoletas).

### Pasos actualizados para Tailwind CSS v4 con Vite

1. **Crear (solo si lo necesitas) `tailwind.config.ts`**
   - En la raíz del proyecto, crea el archivo.
   - Pega el contenido mínimo para personalizar rutas y tema cuando haga falta:
     ```ts
     import type { Config } from "tailwindcss";

     export default {
       content: ["./index.html", "./src/**/*.{js,ts,jsx,tsx}"],
       theme: { extend: {} },
       plugins: [],
     } satisfies Config;
     ```

2. **Activar el plugin oficial en `vite.config.ts` (o `.js`)**
   ```ts
   import { defineConfig } from "vite";
   import react from "@vitejs/plugin-react";
   import tailwindcss from "@tailwindcss/vite";

   export default defineConfig({
     plugins: [react(), tailwindcss()],
   });
   ```

3. **Importar Tailwind en tu hoja de estilos principal (`src/index.css`)**
   - Usa el formato clásico:
     ```css
     @tailwind base;
     @tailwind components;
     @tailwind utilities;
     ```
   - o la nueva sintaxis:
     ```css
     @import "tailwindcss";
     ```


5. Ejecutar el servidor:

```bash
npm run dev
```
