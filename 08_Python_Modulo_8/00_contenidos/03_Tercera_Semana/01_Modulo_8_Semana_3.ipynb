# Clase 5 (Consumo de APIs + Rutas + Tokens)

## 1.1. ¿Qué es una API y por qué la necesitamos en React?

**Idea clave:** React se encarga de la interfaz. La **información real** viene de una API.

Esquema para explicar:


<div align="center">
  <img src="img/API2.png" alt="Descripción" width="80%">
</div>


* **API**: punto de acceso donde otra aplicación (tu frontend) pide o envía información.
* Normalmente usa HTTP con formato **JSON**.
* En un proyecto serio: React no “sabe” de la base de datos; solo habla con la API.

Definición sencilla para clase:

> “Una API es un contrato que dice: si me mandas estas cosas, yo te responderé con estas otras. React solo sabe hablar ese contrato, no cómo está implementado por dentro.”

---


## 1.2. Métodos HTTP y su significado

Explica que hay 4 que usamos el 90% del tiempo:

| Método | Significado en REST              | Ejemplo                    |
| ------ | -------------------------------- | -------------------------- |
| GET    | Obtener recursos                 | Ver lista de tareas        |
| POST   | Crear un recurso nuevo           | Crear nueva tarea          |
| PUT    | Reemplazar/actualizar un recurso | Editar una tarea existente |
| DELETE | Eliminar un recurso              | Borrar una tarea           |

Lo conectas con el mundo real:

> GET: “Dame todas las tareas”
> POST: “Agrega esta tarea nueva”
> PUT: “Actualiza esta tarea con estos datos”
> DELETE: “Elimina esta tarea”

---


## 1.3. ¿Por qué Axios y no solo `fetch`?

**Axios** es una biblioteca que mejora la experiencia con HTTP:

* Maneja JSON automáticamente (no tienes que hacer `res.json()`).
* Permite crear una **instancia** con `baseURL` y headers por defecto.
* Soporta **interceptores** (perfecto para agregar o actualizar tokens).
* Manejo más cómodo de errores.

Esquema:

  <img src="img/AXIOS.png" alt="Descripción" width="80%">
</div>


> “Axios es como un mensajero configurado: ya sabe a qué servidor ir, qué idioma hablar (JSON) y qué credenciales llevar (token). Tú solo le dices qué quieres hacer (GET/POST/PUT/DELETE).”

## 1.4. JWT (JSON Web Token) – idea simple

No necesitas entrar a criptografía, pero sí a lo conceptual:

* Tras hacer **login**, el backend genera un token (una cadena larga).
* Ese token representa que el usuario está autenticado.
* El frontend **guarda el token** (por ejemplo, en `localStorage`).
* En cada llamada a la API se envía:

```http
Authorization: Bearer <TOKEN>
```

El backend revisa:

* ¿Es un token válido?
* ¿No está expirado?
* ¿Este usuario tiene permisos?

Si algo falla → responde con 401 o 403 y el frontend debe reaccionar (redirigir a login, mostrar mensaje).

## 1.6. Estado de UI: `loading`, `error` y renderizado condicional

Cuando llamamos una API, pasan 3 momentos:

1. Antes de llamar: `loading = true`, sin datos aún.
2. Si responde bien: `loading = false`, tenemos datos.
3. Si falla: `loading = false`, dejamos un mensaje en `error`.

En código esto se refleja con:

* Mensaje tipo: *“Cargando tareas…”*
* Mensaje tipo: *“No se pudieron cargar las tareas. Intente de nuevo.”*
* Botones deshabilitados mientras se envían datos.

Esto es **UX profesional**, no solo “funciona”.

## 1.7. Hooks personalizados para consumo de API

En lugar de escribir:

```jsx
useEffect(() => {
  axios.get(...).then(...).catch(...);
}, []);
```

en cada componente, creamos un **hook reutilizable**, por ejemplo `useTasks()`.

Ventajas:

* Código más limpio.
* Cambiar la lógica en un solo lugar.
* Facilitar pruebas.

Estructura conceptual:

```text
Componente UI (Dashboard)
      ▲
      │ consume
      ▼
Hook useTasks (encapsula lógica de Axios, loading, error)
      ▲
      │ consume
      ▼
Axios Client (configurado con token)
```

## 3.2. Configurar Axios con token

📄 `src/api/client.js`


````python
// src/api/client.js
import axios from "axios";

// 1. Creamos una instancia de Axios con una baseURL.
export const api = axios.create({
  baseURL: "http://localhost:3001", // Aquí iría tu API real o mock
  headers: {
    "Content-Type": "application/json",
  },
});

// 2. Interceptor: antes de cada request, agregamos el token si existe.
api.interceptors.request.use((config) => {
  const token = localStorage.getItem("token");
  if (token) {
    config.headers.Authorization = `Bearer ${token}`;
  }
  return config;
});
````


> “A partir de aquí, cualquier llamada que hagamos con `api.get`, `api.post`, etc., llevará automáticamente el token en el header.”

## 3.3. Configurar React Router y la ruta protegida

`src/components/PrivateRoute.jsx`



````python
// src/components/PrivateRoute.jsx

// Importamos 'Navigate' desde react-router-dom, que sirve para redirigir al usuario a otra ruta.
import { Navigate } from "react-router-dom";

// Definimos el componente 'PrivateRoute', que recibe 'children' (el contenido que debe proteger).
export const PrivateRoute = ({ children }) => {
  
  // Obtenemos el token del localStorage. Si el usuario está autenticado, este token debe existir.
  const token = localStorage.getItem("token");
  
  // Si no existe token ⇒ el usuario NO está autenticado.
  if (!token) {
    // Redirige al usuario a la página de login. 'replace' evita que pueda volver atrás con el navegador.
    return <Navigate to="/login" replace />;
  }

  // Si el token existe ⇒ devuelve los componentes hijos (children), permitiendo el acceso a la ruta protegida.
  return children;
};
````

`src/App.jsx`



````python
// src/App.jsx
import { BrowserRouter, Routes, Route, Navigate } from "react-router-dom";
import Login from "./pages/Login";
import TasksList from "./pages/TasksList";
import TaskPage from "./pages/TaskPage";
import { PrivateRoute } from "./components/PrivateRoute";

function App() {
  return (
    <BrowserRouter>
      <Routes>
        {/* Ruta de login */}
        <Route path="/login" element={<Login />} />

        {/* Ruta protegida: listado de tareas */}
        <Route
          path="/tasks"
          element={
            <PrivateRoute>
              <TasksList />
            </PrivateRoute>
          }
        />

        {/* Ruta protegida: crear nueva tarea */}
        <Route
          path="/tasks/new"
          element={
            <PrivateRoute>
              <TaskPage mode="create" />
            </PrivateRoute>
          }
        />

        {/* Ruta protegida: editar tarea */}
        <Route
          path="/tasks/:id"
          element={
            <PrivateRoute>
              <TaskPage mode="edit" />
            </PrivateRoute>
          }
        />

        {/* Cualquier otra ruta redirige a login */}
        <Route path="*" element={<Navigate to="/login" replace />} />
      </Routes>
    </BrowserRouter>
  );
}

export default App;
````

## 3.4. Página de Login con validaciones y token

`src/pages/Login.jsx`



````python
import { useState } from "react";
import { useNavigate } from "react-router-dom";

function Login() {
  const [email, setEmail] = useState("");
  const [password, setPassword] = useState("");
  const [error, setError] = useState("");
  const navigate = useNavigate();

  const handleSubmit = (e) => {
    e.preventDefault();

    // Validación simple
    if (!email.includes("@")) {
      setError("El email no es válido.");
      return;
    }
    if (password.length < 4) {
      setError("La contraseña debe tener al menos 4 caracteres.");
      return;
    }

    // Aquí, en un backend real, haríamos:
    // const res = await api.post('/login', { email, password });
    // const token = res.data.token;

    const fakeToken = "TOKEN_JWT_FICTICIO_123";
    localStorage.setItem("token", fakeToken);

    setError("");
    navigate("/tasks");
  };

  return (
    <div className="min-h-screen flex items-center justify-center bg-slate-100">
      <form
        onSubmit={handleSubmit}
        className="bg-white shadow-md rounded p-6 w-full max-w-sm"
      >
        <h1 className="text-xl font-semibold mb-4 text-center">
          Iniciar sesión
        </h1>

        {error && (
          <p className="text-red-500 text-sm mb-2">
            {error}
          </p>
        )}

        <label className="block mb-2 text-sm font-medium">
          Correo electrónico
        </label>
        <input
          className={`border rounded w-full p-2 mb-3 ${
            error && !email.includes("@") ? "border-red-500" : "border-gray-300"
          }`}
          type="email"
          placeholder="usuario@empresa.com"
          value={email}
          onChange={(e) => setEmail(e.target.value)}
        />

        <label className="block mb-2 text-sm font-medium">
          Contraseña
        </label>
        <input
          className={`border rounded w-full p-2 mb-4 ${
            error && password.length < 4 ? "border-red-500" : "border-gray-300"
          }`}
          type="password"
          placeholder="••••••••"
          value={password}
          onChange={(e) => setPassword(e.target.value)}
        />

        <button
          type="submit"
          className="w-full bg-blue-600 text-white py-2 rounded hover:bg-blue-700 transition"
        >
          Entrar
        </button>
      </form>
    </div>
  );
}

export default Login;
````

## 3.5. Hook `useTasks` para encapsular consumo de API



`src/hooks/useTasks.js`

````python
// src/hooks/useTasks.js
import { useEffect, useState } from "react";
import { api } from "../api/client";

export const useTasks = () => {
  const [tasks, setTasks] = useState([]);
  const [loading, setLoading] = useState(true);
  const [error, setError] = useState("");

  const loadTasks = async () => {
    try {
      setLoading(true);
      setError("");
      const res = await api.get("/tasks");
      setTasks(res.data);
    } catch (err) {
      console.error(err);
      setError("No se pudieron cargar las tareas.");
    } finally {
      setLoading(false);
    }
  };

  useEffect(() => {
    loadTasks();
  }, []);

  return { tasks, loading, error, reload: loadTasks };
};
````

Puedes explicar que:

> “Ahora cualquier componente que use `useTasks()` tiene listo: datos, loading, error y una función `reload`.”

## 3.6. Listado de tareas con botones de acción

`src/pages/TasksList.jsx`



````python
import { Link, useNavigate } from "react-router-dom";
import { useTasks } from "../hooks/useTasks";
import { api } from "../api/client";

function TasksList() {
  const { tasks, loading, error, reload } = useTasks();
  const navigate = useNavigate();

  const handleDelete = async (id) => {
    if (!confirm("¿Seguro que deseas eliminar esta tarea?")) return;
    await api.delete(`/tasks/${id}`);
    reload();
  };

  const handleLogout = () => {
    localStorage.removeItem("token");
    navigate("/login");
  };

  const getStatusBadge = (status) => {
    const styles = {
      Pendiente: "bg-yellow-100 text-yellow-800",
      "En progreso": "bg-blue-100 text-blue-800",
      Completada: "bg-green-100 text-green-800",
    };
    return (
      <span
        className={`px-2 py-1 rounded-full text-xs font-medium ${
          styles[status] || "bg-gray-100 text-gray-800"
        }`}
      >
        {status}
      </span>
    );
  };

  const getPriorityBadge = (priority) => {
    const styles = {
      Alta: "bg-red-100 text-red-800",
      Media: "bg-orange-100 text-orange-800",
      Baja: "bg-gray-100 text-gray-800",
    };
    return (
      <span
        className={`px-2 py-1 rounded-full text-xs font-medium ${
          styles[priority] || "bg-gray-100 text-gray-800"
        }`}
      >
        {priority}
      </span>
    );
  };

  if (loading)
    return (
      <div className="min-h-screen bg-slate-100 flex items-center justify-center">
        <p className="text-lg text-gray-600">Cargando tareas...</p>
      </div>
    );
  if (error)
    return (
      <div className="min-h-screen bg-slate-100 flex items-center justify-center">
        <p className="text-lg text-red-500">{error}</p>
      </div>
    );

  return (
    <div className="min-h-screen bg-slate-100 py-8 px-4">
      <div className="max-w-7xl mx-auto">
        {/* Header */}
        <div className="flex flex-col sm:flex-row justify-between items-start sm:items-center mb-6 gap-4">
          <h1 className="text-3xl font-bold text-gray-800">Panel de tareas</h1>
          <div className="flex gap-3">
            <Link
              to="/tasks/new"
              className="bg-blue-600 text-white px-5 py-2 rounded-lg hover:bg-blue-700 transition shadow-md font-medium"
            >
              Nueva tarea
            </Link>
            <button
              onClick={handleLogout}
              className="bg-gray-600 text-white px-5 py-2 rounded-lg hover:bg-gray-700 transition shadow-md font-medium"
            >
              Cerrar sesión
            </button>
          </div>
        </div>

        {/* Table Container */}
        {tasks.length === 0 ? (
          <div className="bg-white rounded-lg shadow-md p-12 text-center">
            <p className="text-gray-500 text-lg">
              No hay tareas registradas.{" "}
              <Link
                to="/tasks/new"
                className="text-green-600 hover:underline font-medium"
              >
                Crea una nueva tarea
              </Link>
            </p>
          </div>
        ) : (
          <div className="bg-white rounded-lg shadow-md overflow-hidden">
            <div className="overflow-x-auto">
              <table className="w-full">
                <thead className="bg-gray-50 border-b">
                  <tr>
                    <th className="px-6 py-4 text-left text-xs font-semibold text-gray-700 uppercase tracking-wider">
                      Título
                    </th>
                    <th className="px-6 py-4 text-left text-xs font-semibold text-gray-700 uppercase tracking-wider">
                      Estado
                    </th>
                    <th className="px-6 py-4 text-left text-xs font-semibold text-gray-700 uppercase tracking-wider">
                      Prioridad
                    </th>
                    <th className="px-6 py-4 text-left text-xs font-semibold text-gray-700 uppercase tracking-wider">
                      Responsable
                    </th>
                    <th className="px-6 py-4 text-left text-xs font-semibold text-gray-700 uppercase tracking-wider">
                      Vence
                    </th>
                    <th className="px-6 py-4 text-left text-xs font-semibold text-gray-700 uppercase tracking-wider">
                      Acciones
                    </th>
                  </tr>
                </thead>
                <tbody className="bg-white divide-y divide-gray-200">
                  {tasks.map((t) => (
                    <tr
                      key={t.id}
                      className="hover:bg-gray-50 transition-colors"
                    >
                      <td className="px-6 py-4 whitespace-nowrap">
                        <div className="text-sm font-medium text-gray-900">
                          {t.title}
                        </div>
                      </td>
                      <td className="px-6 py-4 whitespace-nowrap">
                        {getStatusBadge(t.status)}
                      </td>
                      <td className="px-6 py-4 whitespace-nowrap">
                        {getPriorityBadge(t.priority)}
                      </td>
                      <td className="px-6 py-4 whitespace-nowrap text-sm text-gray-700">
                        {t.assignedTo}
                      </td>
                      <td className="px-6 py-4 whitespace-nowrap text-sm text-gray-700">
                        {t.dueDate}
                      </td>
                      <td className="px-6 py-4 whitespace-nowrap text-sm">
                        <div className="flex gap-2">
                          <button
                            className="bg-blue-600 text-white px-3 py-1.5 rounded hover:bg-blue-700 transition text-xs font-medium"
                            onClick={() => navigate(`/tasks/${t.id}`)}
                          >
                            Editar
                          </button>
                          <button
                            className="bg-red-600 text-white px-3 py-1.5 rounded hover:bg-red-700 transition text-xs font-medium"
                            onClick={() => handleDelete(t.id)}
                          >
                            Eliminar
                          </button>
                        </div>
                      </td>
                    </tr>
                  ))}
                </tbody>
              </table>
            </div>
          </div>
        )}
      </div>
    </div>
  );
}

export default TasksList;
````

## 3.7. Formulario grande de tarea (crear y editar)

`src/components/TaskForm.jsx`



````python
import { useState } from "react";

const EMPTY_TASK = {
  title: "",
  description: "",
  priority: "Media",
  status: "Pendiente",
  dueDate: "",
  assignedTo: "",
  tags: "",
};

function TaskForm({ initialData, onSubmit, loading }) {
  const [form, setForm] = useState(() => {
    if (initialData) {
      return {
        ...initialData,
        tags: initialData.tags ? initialData.tags.join(", ") : "",
      };
    }
    return EMPTY_TASK;
  });
  const [errors, setErrors] = useState({});

  const validate = () => {
    const newErrors = {};
    if (!form.title.trim()) newErrors.title = "El título es obligatorio.";
    if (!form.description.trim() || form.description.length < 10)
      newErrors.description = "La descripción debe tener al menos 10 caracteres.";
    if (!form.dueDate) newErrors.dueDate = "La fecha de vencimiento es obligatoria.";
    if (!form.assignedTo.trim())
      newErrors.assignedTo = "Debe asignarse un responsable.";
    return newErrors;
  };

  const handleChange = (e) => {
    const { name, value } = e.target;
    setForm((prev) => ({
      ...prev,
      [name]: value,
    }));
  };

  const handleSubmit = (e) => {
    e.preventDefault();
    const v = validate();
    if (Object.keys(v).length > 0) {
      setErrors(v);
      return;
    }
    setErrors({});
    const payload = {
      ...form,
      tags: form.tags
        ? form.tags.split(",").map((t) => t.trim()).filter(Boolean)
        : [],
    };
    onSubmit(payload);
  };

  return (
    <form onSubmit={handleSubmit} className="bg-white shadow rounded p-4 space-y-4">
      {/* Título */}
      <div>
        <label className="block text-sm font-medium mb-1">Título</label>
        <input
          name="title"
          value={form.title}
          onChange={handleChange}
          className={`border rounded w-full p-2 ${
            errors.title ? "border-red-500" : "border-gray-300"
          }`}
          placeholder="Ej: Revisar reporte financiero Q4"
        />
        {errors.title && (
          <p className="text-red-500 text-xs mt-1">{errors.title}</p>
        )}
      </div>

      {/* Descripción */}
      <div>
        <label className="block text-sm font-medium mb-1">Descripción</label>
        <textarea
          name="description"
          value={form.description}
          onChange={handleChange}
          className={`border rounded w-full p-2 min-h-[80px] ${
            errors.description ? "border-red-500" : "border-gray-300"
          }`}
          placeholder="Describe el objetivo, contexto y entregables de la tarea."
        />
        {errors.description && (
          <p className="text-red-500 text-xs mt-1">{errors.description}</p>
        )}
      </div>

      {/* Fila prioridad / estado */}
      <div className="grid grid-cols-1 md:grid-cols-2 gap-4">
        <div>
          <label className="block text-sm font-medium mb-1">Prioridad</label>
          <select
            name="priority"
            value={form.priority}
            onChange={handleChange}
            className="border rounded w-full p-2 border-gray-300"
          >
            <option>Baja</option>
            <option>Media</option>
            <option>Alta</option>
          </select>
        </div>

        <div>
          <label className="block text-sm font-medium mb-1">Estado</label>
          <select
            name="status"
            value={form.status}
            onChange={handleChange}
            className="border rounded w-full p-2 border-gray-300"
          >
            <option>Pendiente</option>
            <option>En progreso</option>
            <option>Completada</option>
          </select>
        </div>
      </div>

      {/* Fecha y responsable */}
      <div className="grid grid-cols-1 md:grid-cols-2 gap-4">
        <div>
          <label className="block text-sm font-medium mb-1">
            Fecha de vencimiento
          </label>
          <input
            type="date"
            name="dueDate"
            value={form.dueDate}
            onChange={handleChange}
            className={`border rounded w-full p-2 ${
              errors.dueDate ? "border-red-500" : "border-gray-300"
            }`}
          />
          {errors.dueDate && (
            <p className="text-red-500 text-xs mt-1">{errors.dueDate}</p>
          )}
        </div>

        <div>
          <label className="block text-sm font-medium mb-1">
            Responsable
          </label>
          <input
            name="assignedTo"
            value={form.assignedTo}
            onChange={handleChange}
            className={`border rounded w-full p-2 ${
              errors.assignedTo ? "border-red-500" : "border-gray-300"
            }`}
            placeholder="Nombre de la persona responsable"
          />
          {errors.assignedTo && (
            <p className="text-red-500 text-xs mt-1">{errors.assignedTo}</p>
          )}
        </div>
      </div>

      {/* Tags */}
      <div>
        <label className="block text-sm font-medium mb-1">
          Tags (separados por coma)
        </label>
        <input
          name="tags"
          value={form.tags}
          onChange={handleChange}
          className="border rounded w-full p-2 border-gray-300"
          placeholder="urgente, contabilidad, cliente X"
        />
      </div>

      <button
        type="submit"
        disabled={loading}
        className={`w-full py-2 rounded text-white ${
          loading
            ? "bg-gray-400 cursor-not-allowed"
            : "bg-blue-600 hover:bg-blue-700"
        }`}
      >
        {loading ? "Guardando..." : "Guardar tarea"}
      </button>
    </form>
  );
}

export default TaskForm;
````

Solo con este formulario tienes material para más de una hora de explicación:

* Validaciones manuales.
* Clases dinámicas con Tailwind.
* Conversión de tags string → array.
* Manejo de estado avanzado.

## 3.8. Página que usa TaskForm para crear y editar

`src/pages/TaskPage.jsx`



````python
import { useEffect, useState } from "react";
import { useParams, useNavigate } from "react-router-dom";
import TaskForm from "../components/TaskForm";
import { api } from "../api/client";

function TaskPage({ mode }) {
  const { id } = useParams();
  const navigate = useNavigate();
  const [initialData, setInitialData] = useState(null);
  const [loading, setLoading] = useState(mode === "edit");
  const [saving, setSaving] = useState(false);

  useEffect(() => {
    if (mode === "edit" && id) {
      const loadTask = async () => {
        try {
          const res = await api.get(`/tasks/${id}`);
          setInitialData(res.data);
        } catch (err) {
          console.error(err);
          alert("No se pudo cargar la tarea.");
          navigate("/tasks");
        } finally {
          setLoading(false);
        }
      };
      loadTask();
    }
  }, [mode, id, navigate]);

  const handleSubmit = async (payload) => {
    try {
      setSaving(true);
      if (mode === "create") {
        await api.post("/tasks", payload);
      } else {
        await api.put(`/tasks/${id}`, payload);
      }
      navigate("/tasks");
    } catch (err) {
      console.error(err);
      alert("Ocurrió un error al guardar la tarea.");
    } finally {
      setSaving(false);
    }
  };

  if (loading) return <p className="p-6">Cargando tarea...</p>;

  return (
    <div className="min-h-screen bg-slate-100 p-6">
      <div className="max-w-2xl mx-auto">
        <h1 className="text-2xl font-semibold mb-4">
          {mode === "create" ? "Nueva tarea" : "Editar tarea"}
        </h1>
        <TaskForm
          key={initialData?.id || "new"}
          initialData={initialData}
          onSubmit={handleSubmit}
          loading={saving}
        />
      </div>
    </div>
  );
}

export default TaskPage;
````

Aquí tienes para explicar:

* Diferencia entre modo `create` y `edit`.
* Uso de `useParams` para tomar `id` de la URL.
* Control de `loading` y `saving`.
