## **Prompt para Cursor AI — Clase 2: Modularización de componentes y refactor con IA**

> **Rol:** Eres un asistente experto en desarrollo frontend con **React 18, Vite 5 y Tailwind CSS v4**, especializado en **refactorización, arquitectura limpia y documentación técnica**.
> Tu tarea es **modularizar y documentar completamente** una interfaz React existente de un sistema de tareas, aplicando los principios de **separación de lógica y presentación**, **reutilización de componentes**, y **diseño responsivo moderno** con TailwindCSS.

---

### **Objetivos de la clase**

1. Separar la lógica y la interfaz (UI) en componentes reutilizables.
2. Mejorar la organización, estructura y mantenibilidad del código.
3. Aplicar refactorización con asistencia de IA para lograr código limpio y documentado.
4. Implementar diseño responsivo básico con TailwindCSS.

---

### **Actividades solicitadas**

Refactoriza el proyecto para que contenga los siguientes componentes dentro de `src/components/`:

| **Componente** | **Función**         | **Descripción técnica**                                                                                          |
| -------------- | ------------------- | ---------------------------------------------------------------------------------------------------------------- |
| `Header.jsx`   | Cabecera principal  | Muestra el título del sistema y un subtítulo o mensaje de bienvenida.                                            |
| `TaskList.jsx` | Contenedor de lista | Renderiza el listado de tareas dinámicamente a partir de un arreglo.                                             |
| `TaskItem.jsx` | Tarea individual    | Componente individual reutilizable para cada tarea. Debe recibir `props` como `title`, `completed` y `onToggle`. |
| `Footer.jsx`   | Pie de página       | Muestra información institucional o de autoría (por ejemplo, “Sistema de Tareas © 2025”).                        |

---

### **Requerimientos técnicos**

1. **React funcional con Hooks (`useState`, `useEffect`)**.
2. **Tailwind CSS v4** para todos los estilos (no CSS externo).
3. Código documentado con comentarios en formato JSDoc.
4. Seguir las normas:

   * Airbnb JavaScript Style Guide.
   * OWASP Secure Coding Practices (sin `eval`, ni `innerHTML`).
   * Clean Code (funciones cortas, nombres claros, sin duplicación).
5. Implementar **diseño responsivo básico** (breakpoints `sm`, `md`, `lg`).
6. Evitar lógica repetida: la IA debe detectar patrones similares y abstraerlos en componentes.

---

### **Asistencia con IA**

La IA debe:

* Identificar elementos repetidos (por ejemplo, tarjetas o filas) y convertirlos en componentes reutilizables.
* Refactorizar JSX y estructura de carpetas para mayor legibilidad.
* Generar comentarios explicativos sobre la responsabilidad de cada componente.
* Agregar documentación inline y un **resumen técnico al final del archivo principal (`App.jsx`)** explicando cómo se modularizó el sistema.

---

### **Documentación esperada**

Cada componente debe incluir:

1. **Cabecera JSDoc** con descripción, props y autor.
2. **Comentarios inline** explicando decisiones clave.
3. Un bloque `/** IA Refactor Summary: … */` al final del archivo con un resumen de la mejora realizada.

Ejemplo de cabecera:

```jsx
/**
 * TaskItem.jsx
 * Componente reutilizable para representar una tarea individual.
 * @param {string} title - Nombre de la tarea.
 * @param {boolean} completed - Estado de la tarea.
 * @param {function} onToggle - Callback para alternar estado.
 * @returns {JSX.Element}
 */
```

---

### **Resultado esperado**

Un proyecto con estructura modular:

```
src/
├── components/
│   ├── Header.jsx
│   ├── TaskList.jsx
│   ├── TaskItem.jsx
│   ├── Footer.jsx
│
├── App.jsx
└── main.jsx
```

**Características del resultado:**

* Código limpio, seguro y mantenible.
* Documentación generada automáticamente.
* Diseño visual equilibrado, moderno y responsivo.
* Ejecutable sin errores con:

  ```bash
  npm run dev
  ```
