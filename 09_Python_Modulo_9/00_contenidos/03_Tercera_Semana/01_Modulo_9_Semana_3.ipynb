## Arquitectura obligatoria del Frontend

Respetar estrictamente la siguiente estructura:

```
frontend/
 ├── src/
 │   ├── api/            # Cliente Axios, interceptores, endpoints
 │   ├── auth/           # Login, guards, sesión, roles
 │   ├── components/     # Componentes reutilizables (UI puro)
 │   ├── pages/          # Vistas completas por caso de uso
 │   ├── layouts/        # Layouts base (Auth / Dashboard)
 │   ├── hooks/          # Hooks personalizados
 │   ├── context/        # Estado global (auth, usuario)
 │   ├── services/       # Lógica de negocio frontend
 │   ├── routes/         # Rutas públicas y protegidas
 │   ├── utils/          # Helpers, validaciones, sanitización
 │   ├── styles/         # Estilos globales
 │   └── main.tsx
 └── package.json
```

### Principios obligatorios

* Separation of Concerns (SoC)
* Single Responsibility Principle (SRP)
* DRY
* Tipado estricto con TypeScript
* Component-Based Architecture
* Flux / State Management Pattern

## Seguridad y Hardening del Frontend (OBLIGATORIO)

Implementar y validar:

### Seguridad

* Autenticación real con backend
* Manejo seguro de JWT / Cookies HttpOnly
* Nunca usar LocalStorage para tokens
* Axios con interceptores para:

  * Autenticación
  * Expiración de sesión
  * Errores centralizados
* Protección de rutas por:

  * Autenticación
  * Rol
* Sanitización de entradas
* Prevención de XSS
* Manejo seguro de errores (no filtrar detalles internos)

### Hardening

* Eliminación de:

  * console.log en producción
  * warnings innecesarios
* Validación estricta de tipos
* Manejo controlado de estados inválidos
* Variables sensibles solo en `.env`
* Configuración por ambiente (dev / prod)

Alineado con:

* OWASP Frontend Security Guidelines

# Inicio de la aplicación

Para levantar y probar todo:

**Backend**

En la raíz del proyecto:

`cd backend`

`pip install -r requirements.txt`

`uvicorn app.main:app --reload`

Queda en `http://127.0.0.1:8000` (Swagger: /docs).

**Frontend**

En otra terminal:

`cd frontend`

`npm install`

`npm run dev`

Queda en `http://localhost:3000`

Configuración mínima

* Backend: .env en la raíz

* Frontend: frontend/.env con: VITE_API_BASE_URL=http://127.0.0.1:8000
