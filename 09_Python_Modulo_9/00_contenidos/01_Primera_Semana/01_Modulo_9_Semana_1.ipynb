## CLASE 1

### Planificación, arquitectura y diseño del proyecto Full Stack

### 1. Contexto general del proyecto

**Descripción del sistema**
El sistema de gestión académica universitaria permitirá:

* Gestión de usuarios con roles (administrador, docente, estudiante)
* Gestión de materias
* Inscripción de estudiantes a materias
* Registro y consulta de calificaciones
* Autenticación y autorización segura
* Reportes académicos básicos

**Principios internacionales aplicados**

* ISO/IEC 25010: calidad del software (mantenibilidad, seguridad, escalabilidad)
* OWASP Top 10: prevención de vulnerabilidades comunes
* Clean Architecture (Robert C. Martin)
* PEP 8 y PEP 257 (estilo y documentación en Python)
* RESTful API Design Guidelines

---


### 2. Arquitectura por capas

**Modelo conceptual**

```
┌──────────────────────┐
│      Frontend        │  React
│  (UI / UX / Views)   │
└─────────▲────────────┘
          │ HTTPS (JSON)
┌─────────┴────────────┐
│        API           │  FastAPI
│ Controllers / Routes │
└─────────▲────────────┘
          │ ORM
┌─────────┴────────────┐
│     Services         │  Lógica de negocio
└─────────▲────────────┘
          │
┌─────────┴────────────┐
│    Persistence       │  SQLAlchemy
│  PostgreSQL Database │
└──────────────────────┘
```

**Responsabilidades por capa**

* **Frontend:** validación básica, experiencia de usuario
* **API / Rutas:** validación de entrada, control de acceso
* **Servicios:** reglas de negocio, transacciones
* **Persistencia:** acceso a datos, integridad referencial

### 3. Diseño de módulos del sistema


#### 3.1 Backend

**Separación por dominios (DDD básico)**

```
backend/
 ├── app/
 │   ├── auth/           # Autenticación y autorización (login, JWT, roles, guards)
 │   ├── users/          # Gestión de usuarios (CRUD, estados, asignación de roles)
 │   ├── subjects/       # Gestión de materias académicas
 │   ├── enrollments/    # Gestión de inscripciones académicas
 │   ├── grades/         # Gestión de calificaciones
 │   ├── core/           # Configuración global, seguridad, base de datos y utilidades
 │   └── main.py         # Punto de entrada de la aplicación FastAPI

```

#### 3.2 Arquitectura del Frontend (React) – Enfoque profesional

El Frontend se diseña siguiendo principios de **arquitectura modular**, **separación de responsabilidades** y **escalabilidad**, alineado con estándares industriales usados en aplicaciones empresariales.

**Principios aplicados**

* **Arquitectura basada en componentes** (Component-Based Architecture)
* **Separación de responsabilidades** (Separation of Concerns – SoC)
* **Principio de responsabilidad única** (Single Responsibility Principle – SRP)
* **Patrón Flux / Patrón de gestión de estado** (Flux / State Management Pattern)
* **Guías de seguridad OWASP para aplicaciones Frontend** (OWASP Frontend Security Guidelines)


---

#### 3.3 Estructura de carpetas del Frontend

```
frontend/
 ├── src/
 │   ├── api/            # Comunicación con backend (Axios / Fetch)
 │   ├── auth/           # Lógica de autenticación y guards
 │   ├── components/     # Componentes reutilizables (UI)
 │   ├── pages/          # Vistas principales (screens)
 │   ├── layouts/        # Layouts base (Dashboard, Auth)
 │   ├── hooks/          # Custom hooks
 │   ├── context/        # Context API / State global
 │   ├── services/       # Lógica de negocio frontend
 │   ├── routes/         # Definición de rutas protegidas
 │   ├── utils/          # Funciones auxiliares
 │   ├── styles/         # Estilos globales
 │   └── main.tsx
 └── package.json
```

---

#### 3.4 Responsabilidades por capa en el Frontend

```
UI (Pages / Components)
        │
        ▼
Hooks / Context (estado)
        │
        ▼
Services (reglas de negocio)
        │
        ▼
API Layer (HTTP)
        │
        ▼
Backend (FastAPI)
```

**Descripción por capa**

* **Pages:** representan casos de uso completos (Login, Gestión de Materias)
* **Components:** reutilizables, sin lógica de negocio
* **Hooks / Context:** manejo de estado y efectos
* **Services:** reglas del dominio del frontend
* **API:** comunicación segura con el backend

---

#### 3.5 Mapeo Frontend ↔ Backend por dominio

| Dominio        | Frontend          | Backend     |
| -------------- | ----------------- | ----------- |
| Usuarios       | pages/users       | users       |
| Materias       | pages/subjects    | subjects    |
| Inscripciones  | pages/enrollments | enrollments |
| Calificaciones | pages/grades      | grades      |
| Auth           | auth/             | auth        |


#### 3.6 Seguridad en el Frontend

Buenas prácticas alineadas a OWASP:

* Nunca almacenar JWT en LocalStorage
* Uso de HttpOnly Cookies (cuando aplique)
* Sanitización de entradas
* Protección de rutas
* Manejo centralizado de errores

### 4. Diseño del Modelo Entidad–Relación (ER)

El modelo Entidad–Relación se diseña siguiendo **normalización en 3FN**, **integridad referencial**, y alineación directa con los casos de uso del sistema universitario.

---

#### 4.1 Entidades y atributos

### **USUARIO**

Representa a cualquier persona que accede al sistema.

* **id_usuario** (PK)
* nombres
* apellidos
* email (UNIQUE)
* password_hash
* estado (activo, inactivo)
* fecha_creacion

---

### **ROL**

Define funciones y permisos dentro del sistema.

* **id_rol** (PK)
* nombre (ADMINISTRADOR, DOCENTE, ESTUDIANTE)
* descripcion

---

### **USUARIO_ROL**

Entidad asociativa que resuelve la relación N:M entre usuarios y roles.

* **id_usuario** (PK, FK → USUARIO)
* **id_rol** (PK, FK → ROL)

---

### **MATERIA**

Representa una asignatura del plan académico.

* **id_materia** (PK)
* nombre
* codigo (UNIQUE)
* creditos
* estado (activa, inactiva)

---

### **PERIODO_ACADEMICO**

Define intervalos oficiales de actividad académica.

* **id_periodo** (PK)
* nombre (ej.: 2025-1)
* fecha_inicio
* fecha_fin
* estado (abierto, cerrado)

---

### **INSCRIPCION**

Entidad transaccional que registra la matrícula de un estudiante en una materia durante un período determinado.

* **id_inscripcion** (PK)
* id_usuario (FK → USUARIO)
* id_materia (FK → MATERIA)
* id_periodo (FK → PERIODO_ACADEMICO)
* fecha_inscripcion
* estado (activa, cancelada)

---

### **CALIFICACION**

Registra evaluaciones académicas asociadas a una inscripción.

> Una inscripción puede generar **múltiples calificaciones**.

* **id_calificacion** (PK)
* id_inscripcion (FK → INSCRIPCION)
* nota
* tipo (parcial, final, recuperación)
* fecha_registro

---

## **4.2 Reglas de integridad**

* Un usuario puede tener múltiples inscripciones en diferentes materias y períodos.
* Una inscripción pertenece a un solo usuario, una sola materia y un solo período académico.
* Una inscripción puede generar múltiples calificaciones.

## **4.3 Relaciones y cardinalidades**

| Relación                   | Cardinalidad | Justificación                                     |
| -------------------------- | ------------ | ------------------------------------------------- |
| Usuario – Rol              | N:M          | Varios usuarios pueden tener varios roles         |
| Usuario – Inscripción      | 1:N          | Un usuario puede tener muchas inscripciones       |
| Materia – Inscripción      | 1:N          | Una materia puede tener muchas inscripciones      |
| Periodo – Inscripción      | 1:N          | Un período puede agrupar muchas inscripciones     |
| Inscripción – Calificación | 1:N          | Una inscripción puede tener muchas calificaciones |
