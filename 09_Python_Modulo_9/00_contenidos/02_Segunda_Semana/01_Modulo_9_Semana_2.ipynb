## 1. Configuración segura de entorno (.env)

Configurar y validar el uso de variables de entorno:

```
APP_ENV=production
APP_JWT_SECRET=**************
APP_COOKIE_SECURE=true
APP_CORS_ORIGINS=https://frontend-produccion.com
DATABASE_URL=postgresql://...
```

Requisitos:

* Usar `python-dotenv`
* Centralizar configuración en `core/config.py`
* Diferenciar claramente perfiles:

  * `development`
  * `production`
* Proteger valores sensibles (nunca hardcodear)


## 2. CORS y perfiles de ambiente

Implementar:

* Configuración de **CORS dinámica** según ambiente
* En `dev`: permitir orígenes locales
* En `prod`: solo dominios explícitos

Ejemplo:

* Dev: `http://localhost:3000`
* Prod: `https://frontend-produccion.com`

Asegurar:

* Métodos permitidos controlados
* Headers mínimos necesarios
* Cookies seguras solo en producción

## 3. Endurecimiento de autorización

Implementar políticas de acceso avanzadas:

### Control por rol

* Administrador: acceso total
* Docente: acceso solo a recursos académicos asignados
* Estudiante: acceso solo a **sus propios datos**

### Control por dueño del recurso (Ownership)

* Un estudiante **solo puede ver/modificar sus propias inscripciones y notas**
* Un docente **solo puede gestionar materias e inscripciones asociadas**
* Validar ownership en:

  * Services
  * Dependencias de seguridad

Aplicar:

* Dependencias personalizadas (`Depends`)
* Validaciones explícitas antes de ejecutar lógica de negocio

## 4. Autenticación con Passlib

Implementar correctamente:

* `passlib` con `bcrypt`
* Hashing y verificación segura de contraseñas
* Políticas diferenciadas por perfil:

  * Docente / Administrador
  * Estudiante
* Prevención de:

  * Timing attacks
  * Reuso inseguro de hashes

Actualizar:

* Registro de usuarios
* Login
* Refresh token (si existe)

## 5. Rutas protegidas por perfil

Refactorizar rutas para:

* Usar dependencias explícitas por rol
* Declarar permisos a nivel de endpoint
* Evitar lógica de autorización dentro de controllers

Ejemplo conceptual:

* `@requires_role("ADMIN")`
* `@requires_owner_or_role("DOCENTE")`

## 6. Documentación técnica – READMEv3.md

Crear **READMEv3.md** que documente:

* Diferencias entre dev y prod
* Seguridad aplicada
* Flujo de autenticación
* Gestión de cookies seguras
* Control por rol y ownership
* Variables de entorno requeridas
* Recomendaciones para despliegue

## 7. Validación final obligatoria

Al finalizar:

1. Revisar todo el backend
2. Validar cumplimiento de:

   * OWASP Top 10
   * Clean Architecture
   * SRP / DRY
3. Refactorizar si es necesario
4. Confirmar que el sistema está:

   * Seguro
   * Escalable
   * Listo para producción

## Resultado esperado

* Backend endurecido
* Configuración segura por ambiente
* Autenticación y autorización robustas
* Documentación clara y profesional
* Proyecto listo para despliegue real
