1) ¿Dónde se define la conexión a la base de datos?
Respuesta: En backend/app/core/database.py, usando create_engine y sessionmaker.
2) ¿Qué clase base se utiliza para los modelos ORM?
Respuesta: Base, definida en backend/app/core/database.py como DeclarativeBase.
3) ¿Qué archivo concentra el manejo centralizado de errores?
Respuesta: backend/app/core/error_handlers.py.
4) ¿Qué excepción se usa cuando no existe un recurso?
Respuesta: NotFoundError en backend/app/core/exceptions.py.
5) ¿Dónde se realiza el hashing de contraseñas?
Respuesta: En backend/app/core/security.py con passlib.
6) ¿En qué archivo se declara la tabla de relación N:M entre usuarios y roles?
Respuesta: En backend/app/roles/models.py (user_roles).
7) ¿Qué modelo contiene el campo hashed_password?
Respuesta: User en backend/app/users/models.py.
8) ¿Qué esquema de Pydantic contiene el campo password?
Respuesta: UserCreate en backend/app/users/schemas.py.
9) ¿Qué esquema de respuesta omite el hash de contraseña?
Respuesta: UserResponse en backend/app/users/schemas.py.
10) ¿Dónde se valida que el email sea único al crear un usuario?
Respuesta: En create_user de backend/app/users/services.py.
11) ¿Dónde se define el endpoint POST /users?
Respuesta: En backend/app/users/routes.py.
12) ¿Cómo se obtiene la sesión de BD en las rutas?
Respuesta: Con Depends(get_db) desde backend/app/core/database.py.
13) ¿Qué archivo registra todos los routers de la API?
Respuesta: backend/app/main.py.
14) ¿Qué endpoint expone el health check?
Respuesta: GET /health en backend/app/main.py.
15) ¿Qué entidad representa las materias?
Respuesta: Subject en backend/app/subjects/models.py.
16) ¿Dónde se controla que el código de materia sea único?
Respuesta: En create_subject de backend/app/subjects/services.py.
17) ¿Qué entidad representa los periodos académicos?
Respuesta: AcademicPeriod en backend/app/periods/models.py.
18) ¿Dónde se valida que end_date sea mayor o igual a start_date?
Respuesta: En PeriodCreate y PeriodUpdate de backend/app/periods/schemas.py.
19) ¿Qué entidad representa la inscripción de un estudiante a una materia?
Respuesta: Enrollment en backend/app/enrollments/models.py.
20) ¿Qué restricción evita inscripciones duplicadas?
Respuesta: UniqueConstraint("user_id", "subject_id", "period_id") en Enrollment.
21) ¿Dónde se verifica la existencia de usuario, materia y periodo antes de inscribir?
Respuesta: En _ensure_entities de backend/app/enrollments/services.py.
22) ¿Qué enum define el estado de la inscripción?
Respuesta: EnrollmentStatus en backend/app/enrollments/models.py.
23) ¿Qué entidad representa las calificaciones?
Respuesta: Grade en backend/app/grades/models.py.
24) ¿Dónde se valida el rango de la calificación (0–100)?
Respuesta: En GradeCreate y GradeUpdate de backend/app/grades/schemas.py.
25) ¿En qué archivo se define el CRUD de calificaciones?
Respuesta: En backend/app/grades/services.py.
26) ¿Qué capa contiene la lógica de negocio?
Respuesta: Los archivos services.py de cada dominio.
27) ¿Qué capa contiene los endpoints?
Respuesta: Los archivos routes.py de cada dominio.
28) ¿Qué capa contiene los modelos ORM?
Respuesta: Los archivos models.py de cada dominio.
29) ¿Qué archivo lista las dependencias del backend?
Respuesta: backend/requirements.txt.
30) ¿Dónde está documentada la estructura y requisitos básicos?
Respuesta: En README.md en la raíz del proyecto.
