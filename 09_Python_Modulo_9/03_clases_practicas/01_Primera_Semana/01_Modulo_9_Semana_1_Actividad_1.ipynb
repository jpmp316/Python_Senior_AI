## Actividades complementarias 

**Modalidad:** Individual
**Herramienta:** ChatGPT (con prompts controlados)

---

## Actividad 1 – Comprensión del problema y alcance del sistema

**Tiempo estimado:** 15 minutos

### Objetivo

Asegurar que el estudiante comprenda el **contexto funcional y técnico** del sistema universitario antes de diseñar o programar.

### Instrucciones para el estudiante

Usa ChatGPT para validar y ampliar tu comprensión del sistema.

### Prompt sugerido

```text
Actúa como un analista de sistemas senior.
A partir de este contexto:

"Sistema de gestión académica universitaria con usuarios, roles,
materias, inscripciones, calificaciones y autenticación segura"

1. Enumera los actores del sistema.
2. Describe los principales casos de uso por actor.
3. Identifica qué funcionalidades NO deberían incluirse en una primera versión (MVP).
```

### Evidencia a entregar

* Lista de actores
* Lista de casos de uso
* Justificación del alcance del MVP

## Actividad 2 – Arquitectura por capas aplicada al proyecto

**Tiempo estimado:** 15–20 minutos

### Objetivo

Conectar la **arquitectura teórica por capas** con el proyecto real Full Stack.

### Instrucciones

Solicita a ChatGPT que traduzca la arquitectura conceptual a decisiones técnicas concretas.

### Prompt sugerido

```text
Eres un arquitecto de software.
Explica cómo se aplica una arquitectura por capas
(Frontend, Backend, Base de Datos)
en un sistema universitario Full Stack con:

- React
- FastAPI
- PostgreSQL

Para cada capa indica:
- Responsabilidades
- Qué NO debe hacer esa capa
- Riesgos comunes si se rompe la separación de capas
```

### Evidencia a entregar

* Tabla o esquema por capas
* 1 error común por capa explicado con palabras propias

---

## Actividad 3 – Normas internacionales y buenas prácticas

**Tiempo estimado:** 15 minutos

### Objetivo

Desarrollar criterio profesional sobre **por qué** se usan normas y no solo **cuáles** se usan.

### Prompt sugerido

```text
Actúa como revisor de calidad de software.
Explica de forma clara y aplicada al proyecto universitario:

1. Cómo ISO/IEC 25010 impacta la mantenibilidad y seguridad del sistema
2. Por qué OWASP Top 10 es crítico en una API académica
3. Qué problemas reales se evitan usando Clean Architecture
```

### Evidencia a entregar

* Resumen de máximo 1 página
* Al menos un ejemplo concreto por norma

---

## Actividad 4 – Uso responsable de IA en el proyecto

**Tiempo estimado:** 10–15 minutos

### Objetivo

Concientizar sobre el **rol de la IA como asistente**, no como sustituto del criterio profesional.

### Prompt sugerido

```text
Eres un profesor universitario de ingeniería de software.
Responde:

1. ¿En qué partes del proyecto Full Stack es válido usar ChatGPT?
2. ¿En qué partes NO debería usarse?
3. ¿Qué riesgos existen si se usa IA sin criterio técnico?
```

### Evidencia a entregar

* Lista de usos permitidos
* Lista de usos no permitidos
* Reflexión personal (mínimo 5 líneas)

---

## Actividad 5 – Cierre reflexivo (metacognición)

**Tiempo estimado:** 5–10 minutos

### Pregunta de reflexión (sin ChatGPT)

> ¿Por qué la planificación y la arquitectura son más importantes que escribir código rápido en un proyecto profesional?
