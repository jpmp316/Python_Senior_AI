# Validación avanzada de APIs
---

## BLOQUE 1 – Lectura técnica crítica del README_TEST

### Objetivo

Que el estudiante **entienda el porqué de cada prueba**, no solo el cómo.

### Actividad

Luego de desarrollar el README_TEST **responder por escrito**:

1. ¿Por qué el **orden de pruebas es obligatorio**?
2. ¿Qué dependencias existen entre:

   * usuarios
   * materias
   * periodos
   * inscripciones?
3. ¿Qué pasaría si se intenta:

   * crear una inscripción sin periodo
   * crear una nota sin inscripción?

**Producto esperado**: respuestas técnicas, no descriptivas.

---

## BLOQUE 2 – Pruebas negativas y edge cases 

### Actividad guiada

Los estudiantes deben **inventar y ejecutar** pruebas como:

### Autenticación

* Login con email inexistente
* Login con password correcto pero usuario inactivo
* Login sin enviar body
* Login con campos adicionales (inyección leve)

### Usuarios

* Crear usuario sin role_ids
* Crear usuario con role_id inexistente
* Crear usuario con email inválido
* Intentar reactivar usuario eliminado

### Materias

* Crear materia con credits = 0
* Crear materia con código duplicado
* Actualizar materia inactiva

### Inscripciones

* Inscribir dos veces al mismo estudiante en la misma materia y periodo
* Estudiante intenta inscribir a otro estudiante
* Inscribir con materia inactiva

### Preguntas guía

* ¿Qué código HTTP devuelve?
* ¿El mensaje es claro?
* ¿Es seguro (no revela info interna)?

**Producto esperado**: tabla de resultados.

## BLOQUE 3 – Análisis de roles y ownership

### Objetivo

Validar que **la autorización no es solo JWT**, sino **reglas de negocio**.

### Actividad

Cada grupo toma **un rol**:

* Administrador
* Docente
* Estudiante

Y responde:

| Endpoint | ¿Debería permitir? | ¿Permite realmente? | Riesgo si falla |
| -------- | ------------------ | ------------------- | --------------- |

Ejemplos:

* Estudiante → GET /users
* Docente → DELETE /subjects/{id}
* Estudiante → PUT /grades/{id}

Discusión:

> ¿Dónde debería estar esta validación: ruta, dependencia o servicio?

## BLOQUE 4 – Evaluación de calidad de API


### Checklist de evaluación

Cada estudiante debe evaluar:

| Criterio                    | Cumple | No cumple | Observación |
| --------------------------- | ------ | --------- | ----------- |
| Uso correcto de HTTP verbs  | ⬜      | ⬜         |             |
| Status codes coherentes     | ⬜      | ⬜         |             |
| Nombres de endpoints claros | ⬜      | ⬜         |             |
| Responses consistentes      | ⬜      | ⬜         |             |
| Manejo de errores uniforme  | ⬜      | ⬜         |             |

Preguntas clave:

* ¿Dónde hay sobrecarga de lógica en endpoints?
* ¿Algún endpoint viola SRP?
* ¿Algún endpoint hace demasiado?
